# Introducción
En este primer trabajo se pide diseñar un generador de señales con el objetivo de poder simularlas. Para eso, se programará una función que genere señales senoidales y que permita parametrizar:

-*La amplitud máxima de la senoidal (volts)*


-*Su valor medio (volts)*


-*La frecuencia (Hz)*


-*La fase (radianes)*


-*La cantidad de muestras digitalizada por el ADC (# muestras)*


-*La frecuencia de muestreo del ADC.*



La función debería admitir se llamada de la siguiente manera: **tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)**

# Desarollo

```python
# Importación de las ibrerías de jupyter
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

### Definición de Parámetros

```python
vmax = 1 #Es la amplitud máxima de la senoidal (volts)
dc = 0 #El valor medio o valor dc (volts)
ff = 1 #La frecuencia fundamental de la onda (Hz)
ph = 0 #la fase en radianes
N = 1000 #la cantidad de muestras digitalizada por el ADC
fs = 1000 #la frecuencia de muestreo del ADC
ts = 1/fs #Tiempo entre cada muestra
```

La frecuencia de muestreo (fs) indica cuántas muestras se toman por segundo. Su inversa (1/fs) representa el tiempo entre cada muestra. Esta frecuencia está determinada por el conversor analógico-digital (ADC) y define el tipo de muestreo que se puede realizar. Es una variable limitante, ya que limita la resolución temporal y la capacidad de reconstrucción de la señal.

### Definición de la función

```python
def mi_funcion_sen(vmax, dc, ff, ph, N, fs):
    #tt es el vector de instantes de tiempo en los que se toman las muestras de la señal
    tt = np.linspace(0, (N-1)*ts, num=N) #Vector de N posiciones
    # Argumento del seno: 2π*frecuencia*tiempo + fase
    arg=2*np.pi*ff*tt + ph
    #Señal senoidal
    xx = vmax * np.sin(arg) + dc #xx guarda los valores de la señal en esos tiempos.
    return tt, xx  
    
```

## Señal Senoidal

```python
# Llamo a la función usando las variables definidas antes
tt, xx = mi_funcion_sen(vmax, dc, ff, ph, N, fs)
```

## Gráfico la señal con ff= 1Hz

```python
plt.plot(tt, xx)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señal Senoidal")
plt.grid()
plt.show()
```
