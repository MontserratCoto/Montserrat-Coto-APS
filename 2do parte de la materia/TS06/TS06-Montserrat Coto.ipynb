## Introducción


En este trabajo práctico analizaremos cuatro sistemas representados por ecuaciones en diferencias, que corresponden a distintos filtros digitales. El objetivo principal es hallar la función de transferencia:

$$ T(z) = \frac{Y(z)}{X(z)} $$

de cada sistema en el dominio de la Transformada Z, y a partir de ahí calcular y graficar su respuesta en frecuencia, tanto en módulo como en fase.

El propósito de esta actividad es tener un primer acercamiento al concepto de la transformada Z y comenzar a comprender los sistemas en tiempo discreto, particularmente en cómo se comportan sus filtros al analizar la respuesta en frecuencia desde el dominio z. La transformada Z permite trabajar con señales y sistemas discretos de manera análoga a como la transformada de Laplace lo hace para señales continuas, aunque con diferencias clave, como el criterio de estabilidad: en este caso, los polos deben encontrarse dentro del círculo unitario y no en el semiplano izquierdo como en el caso continuo.

En este trabajo se calcularán las transformadas Z directamente desde las ecuaciones en diferencias. Esta primera aproximación permite visualizar si un filtro es FIR (Finite Impulse Response), es decir, si depende únicamente de las entradas, sin recurrencia. Además, se puede anticipar si se trata de un filtro pasa bajos, pasa altos u otro tipo, lo cual es útil para interpretar su función antes de simular.

Un aspecto central del análisis será interpretar cómo los polos y ceros afectan tanto la magnitud como la fase de la respuesta en frecuencia. También se prestará atención a fenómenos particulares como las caídas en la magnitud (zonas donde la señal se atenúa) y las discontinuidades en la fase, provocadas por la ubicación de los ceros respecto del círculo unitario.

$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n)
$$

#### Transformada Z

Aplicamos la transformada Z a ambos lados:

$$
Y(z) = X(z) \cdot \left( z^{-3} + z^{-2} + z^{-1} + 1 \right)
$$

Entonces, la función de transferencia es:

$$
T(z) = \frac{Y(z)}{X(z)} = \frac{z^{-3} + z^{-2} + z^{-1} + 1}{1} = \frac{1 + z + z^2 + z^3}{z^3}
$$

#### Paso al dominio de la frecuencia

Para obtener la respuesta en frecuencia, evaluamos sobre la circunferencia unidad:
 
$$
z = e^{j\omega}
$$

Entonces:

$$
T(e^{j\omega}) = \frac{1 + e^{j\omega} + e^{2j\omega} + e^{3j\omega}}{e^{3j\omega}}
$$

#### Factor común (punto medio = 3/2)

Sacamos factor común el punto medio del exponente:

$$
T(e^{j\omega}) = \frac{e^{\frac{3}{2}j\omega} \left( e^{-\frac{3}{2}j\omega} + e^{-\frac{1}{2}j\omega} + e^{\frac{1}{2}j\omega} + e^{\frac{3}{2}j\omega} \right)}{e^{3j\omega}} = e^{-\frac{3}{2}j\omega} \left( e^{-\frac{3}{2}j\omega} + e^{-\frac{1}{2}j\omega} + e^{\frac{1}{2}j\omega} + e^{\frac{3}{2}j\omega} \right)
$$

#### Aplicando identidades trigonométricas

Aplicamos identidades de Euler:

$$
T(e^{j\omega}) = e^{-\frac{3}{2}j\omega} \left( 2\cos\left(\frac{3}{2}\omega\right) + 2\cos\left(\frac{1}{2}\omega\right) \right)
$$

## C. Validación de la respuesta en frecuencia de todos los sistemas con Numpy

```python
##### Definición de Parámetros
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def plot_freq_response(num, den, title):
    zeros, poles, _ = signal.tf2zpk(num, den)
    w, h = signal.freqz(num, den, worN=8000)

    fig, axs = plt.subplots(1, 3, figsize=(18, 5))

    # Mapa de polos y ceros
    axs[0].plot(np.real(zeros), np.imag(zeros), 'o', label='Ceros', markersize=10)
    axs[0].plot(np.real(poles), np.imag(poles), 'x', label='Polos', markersize=12)
    axs[0].axhline(0, color='gray', lw=1)
    axs[0].axvline(0, color='gray', lw=1)
    axs[0].set_title("Mapa de polos y ceros")
    axs[0].set_xlabel(r'Re{z}')
    axs[0].set_ylabel(r'Im{z}')
    axs[0].legend()

    # Dibujar círculo unidad
    circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='dashed')
    axs[0].add_artist(circle)
    axs[0].set_xlim(-1.5, 1.5)
    axs[0].set_ylim(-1.5, 1.5)
    axs[0].set_aspect('equal', adjustable='box')

    # Respuesta en módulo
    axs[1].plot(w, np.abs(h), 'b')
    axs[1].set_title("Respuesta en módulo")
    axs[1].set_xlabel(r'Frecuencia $\omega$ [rad/muestra]')
    axs[1].set_ylabel(r'$|H(e^{j\omega})|$')
    axs[1].grid(True)
    axs[1].set_xticks([0, np.pi/2, np.pi])
    axs[1].set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])

    # Respuesta en fase
    axs[2].plot(w, np.angle(h), 'r')
    axs[2].set_title("Respuesta en fase")
    axs[2].set_xlabel(r'Frecuencia $\omega$ [rad/muestra]')
    axs[2].set_ylabel(r'$\angle H(e^{j\omega})$ [rad]')
    axs[2].grid(True)
    axs[2].set_xticks([0, np.pi/2, np.pi])
    axs[2].set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])

    # Texto con cantidad de ceros y polos
    n_zeros = len(zeros)
    n_poles = len(poles)
    textstr = f'Ceros: {n_zeros}\nPolos: {n_poles}'
    axs[0].text(0.95, 0.95, textstr, transform=axs[0].transAxes,
                verticalalignment='top', horizontalalignment='right',
                bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.7))

    fig.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

```

```python
num = [1, 1, 1, 1]
den = [1, 0, 0, 0]  # z^3 -> polos en cero

plot_freq_response(num, den, r"Ítem a) $H_a(e^{j\omega}) = \frac{1 + z + z^2 + z^3}{z^3}$")
```

```python
# Ítem b)
num_b = [1, 1, 1, 1, 1]  # 1 + z + z^2 + z^3 + z^4
den_b = [1, 0, 0, 0, 0]  # z^4

plot_freq_response(num_b, den_b, r"Ítem b) $H_b(e^{j\omega}) = \frac{1 + z + z^2 + z^3 + z^4}{z^4}$")
```

```python
# Ítem c)
num_c = [1, -1]  # 1 - z^-1
den_c = [1]      # denominador 1

plot_freq_response(num_c, den_c, r"Ítem c) $H_c(e^{j\omega}) = 1 - z^{-1}$")
```

```python
# Ítem d)
num_d = [1, 0, -1]  # 1 - z^-2
den_d = [1]         # denominador 1

plot_freq_response(num_d, den_d, r"Ítem d) $H_d(e^{j\omega}) = 1 - z^{-2}$")
```

# Análisis  

# 1. Funciones de transferencia y tipo de filtro

Los sistemas analizados son filtros **FIR** (Finite Impulse Response), caracterizados porque la salida **y(n)** depende solo de una combinación lineal de las entradas actuales y pasadas, sin realimentación (es decir, no depende de salidas anteriores). Esto garantiza estabilidad y una respuesta finita en el tiempo.

---
### Ítem a) y(n) = x(n) + x(n-1) + x(n-2) + x(n-3)

La salida es la suma de cuatro muestras consecutivas de entrada. Esto corresponde a un filtro **pasa bajos**, ya que atenúa las altas frecuencias y mantiene las bajas frecuencias, suavizando la señal.

---

### Ítem b) y(n) = x(n) + x(n-1) + x(n-2) + x(n-3) + x(n-4)

Similar al caso anterior, pero sumando cinco muestras consecutivas. Esto refuerza la naturaleza de filtro pasa bajos, aumentando el efecto.

---

### Ítem c) y(n) = x(n) - x(n-1)

En este casoo, la salida es la diferencia entre la entrada actual y la anterior. Este filtro actúa como un **pasa altos**, porque atenúa las componentes de baja frecuencia y deja pasar las altas frecuencias, es decir, resalta los cambios bruscos o transiciones rápidas en la señal.

---

### Ítem d) y(n) = x(n) - x(n-2)

Se resta la entrada con un retraso de dos muestras. Esto genera un filtro con un espectro de rechazo, actuando como un filtro **pasa banda**. Elimina componentes periódicas específicas.

## 2-3) Análisis y validación de la respuesta en frecuencia

Mediante la simulación en Python, se obtuvieron los gráficos de módulo y fase de la respuesta en frecuencia para cada uno de los sistemas. A partir de estos gráficos y del análisis del diagrama de polos y ceros, se pueden destacar las siguientes observaciones:

###  Diagrama de ceros y polos:

- En los sistemas *a)* y *b)* se observan *polos en el origen*, lo cual representa un retraso en el tiempo. Los ceros se distribuyen dentro del círculo unidad o sobre él.
- En los sistemas *c)* y *d)* no hay polos, únicamente ceros ubicados en puntos específicos del círculo unidad.

Todos los sistemas tienen ceros en el plano z y sus polos en el origen o ausentes, por lo que se confirma que son *filtros FIR estables*, ya que sus ceros no se encuentran fuera del círculo unidad.

### Diagrama de fase:

A diferencia del análisis en el plano s, donde se observa el comportamiento en torno al origen, el diagrama de fase muestra la evolución del ángulo de la respuesta sobre la circunferencia de módulo unitario.

Se observa que cuando la frecuencia de evaluación se acerca a la posición de un cero ubicado sobre el círculo unidad, se produce un salto abrupto de π (o -π) en la fase. Esta *discontinuidad* refleja el cambio brusco en la contribución de fase que aporta dicho cero. En cambio, los *polos en el origen* no producen estos saltos abruptos, ya que su contribución en la fase es constante (aportan +1 en magnitud y un cambio de fase suave).



### Validación por función

- En el sistema **a)**, con solo cuatro muestras sumadas, se identifican ceros de transmisión en $\omega = \frac{\pi}{2}$ y $\omega = \pi$. Esto significa que el sistema anula componentes en la mitad de banda y en la frecuencia de Nyquist, lo que refuerza su comportamiento *pasa bajos*.

  La caída de la magnitud en esas frecuencias coincide con la ubicación de los ceros sobre el círculo unidad en el plano $z$, lo que genera zonas de atenuación en el gráfico de módulo.  
  En el gráfico de **fase**, se observa una evolución escalonada, con saltos de $\pm \pi$. Esto ocurre porque, al evaluarse la frecuencia cerca de un cero ubicado sobre el círculo unidad, se produce una variación muy abrupta en la fase. Cada cruce cercano a un cero genera un cambio súbito en la contribución angular.

- En el sistema **b)**, al sumar cinco muestras consecutivas, aparecen más ceros distribuidos sobre el círculo unidad, generando una mayor atenuación en frecuencias medias y altas. El primer cero de transmisión aparece antes que en el sistema a), y hay otro en $\omega = \pi$.

  Esto refuerza su carácter de *filtro pasa bajos*, con una transición más suave en el módulo y mayor rechazo de altas frecuencias.  
  El gráfico de fase presenta más saltos que en el sistema a), debido a la presencia de más ceros sobre el círculo unidad. Cada uno de estos ceros aporta un cambio brusco en la fase cuando la frecuencia se aproxima a su ubicación angular, lo que resulta en una evolución por tramos con discontinuidades de $\pm \pi$ en el gráfico. 

- En el sistema **c)**, con la diferencia simple entre la muestra actual y la anterior, se identifica un cero en $\omega = 0$, es decir, se anula la componente de continua. Esto define claramente un *filtro pasa altos*, ya que deja pasar las componentes de alta frecuencia y atenúa las bajas.

  En el gráfico de módulo se observa una atenuación máxima en 0 y una ganancia creciente hacia $\pi$.  
  El gráfico de fase muestra un salto en $\omega = 0$, que **no debería aparecer** desde el punto de vista teórico. En este caso, la fase es una función lineal continua, pero el salto visual se produce porque la fase calculada comienza en $\frac{\pi}{2}$ y decrece, cruzando el límite del intervalo $[-\pi, \pi]$, lo cual genera un corte automático en el gráfico.

- En el sistema **d)**, donde se restan las entradas con un desfase de 2 muestras, se identifican ceros en $\omega = 0$ y $\omega = \pi$, es decir, en las frecuencias de continua y Nyquist. El filtro actúa como un *pasa banda*, ya que atenúa las frecuencias bajas y altas, mientras que deja pasar las medias (por ejemplo, en $\omega = \frac{\pi}{2}$).

  En el gráfico de módulo se observa un máximo en $\omega = \frac{\pi}{2}$, lo cual confirma el realce de las frecuencias medias.  
  El gráfico de fase muestra una variación lineal, sin saltos abruptos. Este comportamiento se debe a la estructura de la función de transferencia, en particular al término $e^{-j\omega}$. El  filtro no tiene coeficientes simétricos, (son [1, 0, -1]), por lo que la linealidad de fase proviene del desfase constante generado por los ceros complejos conjugados.
