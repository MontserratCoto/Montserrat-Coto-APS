# Introducción

En esta tarea semanal (TS08), el objetivo es analizar y aplicar técnicas **no lineales** para la estimación y substracción de la señal de movimiento de línea base $b(n)$, con la finalidad de obtener una señal ECG limpia y sin interferencias, representada por $\hat{x}$, que se define como: $\hat{x} = s - \hat{b}$ donde $s$ es la señal ECG registrada que contiene interferencias, y $\hat{b}$ es la estimación del movimiento de línea base. 

Se estudiarán dos métodos principales para estimar esta línea base:

1. **Filtro de mediana:** Esta técnica se basa en la aplicación secuencial de filtros de mediana con ventanas de 200 ms y 600 ms. La ventaja del filtro de mediana es su capacidad para eliminar componentes de baja frecuencia y ruidos impulsivos sin distorsionar la morfología de la señal, preservando la integridad de los latidos y otras características importantes del ECG.

2. **Interpolación mediante splines cúbicos:** En este método se utilizan las detecciones de latidos QRS proporcionadas para identificar el segmento PQ de la señal, que es considerado un intervalo isoeléctrico en el que la actividad eléctrica cardíaca es mínima o nula. A partir de estos puntos, se realiza una interpolación spline cúbica para estimar la línea base, garantizando que esta estimación se muestree a la misma frecuencia que la señal original. Esto permite obtener una línea base suave y continua que se adapta a las variaciones lentas de la señal.

Además, se aplicará un **filtro adaptado (matched filter)** para diseñar un detector de latidos QRS basado en un patrón de referencia. Este método es una técnica de detección que maximiza la relación señal-ruido para detectar eventos específicos dentro de señales ruidosas. Se explicará conceptualmente cómo funciona el filtro adaptado, se implementará para detectar latidos y se evaluará su desempeño en comparación con las detecciones reales mediante métricas cuantitativas, como la sensibilidad y el valor predictivo positivo.

##### Definición de Parámetros

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from scipy.signal import medfilt
from scipy.interpolate import CubicSpline
from scipy.signal import find_peaks
from scipy.signal import convolve

plt.close('all')

#############################
#%% Carga de la señal ECG
#############################

fs_ecg = 1000  # Hz
mat_struct = sio.loadmat('./ECG_TP4.mat')
ecg_one_lead = mat_struct['ecg_lead'].flatten()


# Variables del archivo
qrs_pattern = mat_struct['qrs_pattern1'].flatten()
heartbeat_normal = mat_struct['heartbeat_pattern1'].flatten()
heartbeat_ventricular = mat_struct['heartbeat_pattern2'].flatten()
qrs_detections = mat_struct['qrs_detections'].flatten()

# Normalización
ecg_one_lead = ecg_one_lead / np.std(ecg_one_lead)
heartbeat_normal = heartbeat_normal / np.std(heartbeat_normal)
heartbeat_ventricular = heartbeat_ventricular / np.std(heartbeat_ventricular)
qrs_pattern = qrs_pattern / np.std(qrs_pattern)

# Tiempo total señal completa
t_total = np.arange(len(ecg_one_lead)) / fs_ecg
```

## 1) Filtro de mediana

```python
#############################################################################################################################################
#%% Filtro de mediana doble (no lineal) sobre toda la señal
#############################################################################################################################################

win1 = 200  # 200 ms
win2 = 600  # 600 ms

# Asegurar ventanas impares (kernel_size en muestras)
if win1 % 2 == 0: win1 += 1
if win2 % 2 == 0: win2 += 1

# Aplicar filtro mediana doble a la señal completa
mediana1 = medfilt(ecg_one_lead, kernel_size=win1)
mediana2 = medfilt(mediana1, kernel_size=win2)

# Señal filtrada = señal original - línea base estimada
ecg_filtrada = ecg_one_lead - mediana2
```

## 2) Interpolación mediante splines cúbicos 

```python
#############################################################################################################################################
#%% Filtro de Spline Cubico (no lineal) sobre toda la señal
#############################################################################################################################################

N_ecg = len(ecg_one_lead)
t_ecg = np.arange(N_ecg) / fs_ecg  # eje de tiempo en segundos


def estimate_baseline_spline(ecg_signal, qrs_positions, fs=1000, n0_ms=100, window_ms=20):
    """
    Estima la línea de base de una señal ECG usando spline cúbica,
    a partir de puntos en el segmento PQ (aprox. n0_ms antes del QRS).
    El valor se promedia en una ventana de window_ms alrededor del punto.
    
    Parámetros:
    - ecg_signal: señal ECG 1D
    - qrs_positions: índices donde se detectan los QRS (enteros)
    - fs: frecuencia de muestreo (Hz)
    - n0_ms: tiempo en ms para retroceder desde QRS (segmento PQ)
    - window_ms: ventana de promedio en ms
    
    Retorna:
    - baseline: línea de base interpolada (vector del mismo largo que ecg_signal)
    - baseline_times: tiempos de los puntos fiduciales usados para la spline
    - baseline_values: valores promedio en esos puntos
    """
    
    n0 = int(n0_ms * fs / 1000)          # conversión ms a muestras
    window = int(window_ms * fs / 1000)  # ventana en muestras
    
    t_ecg = np.arange(len(ecg_signal)) / fs
    baseline_times = []
    baseline_values = []

    for qrs in qrs_positions:
        idx = qrs - n0
        if idx - window//2 < 0 or idx + window//2 >= len(ecg_signal):
            continue  # evitamos salirnos del vector

        window_data = ecg_signal[idx - window//2 : idx + window//2]
        baseline_times.append(t_ecg[idx])
        baseline_values.append(np.mean(window_data))
        
    spline_func = CubicSpline(baseline_times, baseline_values)
    baseline = spline_func(t_ecg)
    return baseline, baseline_times, baseline_values

# Estimamos línea de base usando spline cúbica
baseline_full, base_times, base_vals = estimate_baseline_spline(ecg_one_lead, qrs_detections, fs=fs_ecg, n0_ms=100, window_ms=20)
ecg_clean_full = ecg_one_lead - baseline_full
```

### Resultado de las variables: $\hat{x}$ , $s$  y  $\hat{b}$ para Spline Cubico

```python
print("\nSpline cúbico - primeros valores de x̂ (filtrada), s (original) y b̂ (línea base):\n")
print(f"{'Índice':>6} | {'x̂ (Filtrada)':^15} | {'s (Original)':^15} | {'b̂ (Línea base)':^18}")
print("-" * 66)
for i in range(20):
    print(f"{i:6d} | {ecg_clean_full[i]:15.6f} | {ecg_one_lead[i]:15.6f} | {baseline_full[i]:18.6f}")
```

## 3) Filtro adaptado (matched filter)

```python
# Centrar señal y patrón para eliminar offset
ecg_centered = ecg_one_lead - np.mean(ecg_one_lead)
pattern_centered = qrs_pattern[::-1] - np.mean(qrs_pattern)

#  Filtrado por convolución (matched filter)
filtered_signal = np.convolve(ecg_centered, pattern_centered, mode='same')

#  Detección de picos en señal filtrada 
height_threshold = np.max(filtered_signal) * 0.4
distance_threshold = int(fs_ecg * 0.3)
peaks, _ = find_peaks(filtered_signal, height=height_threshold, distance=distance_threshold)

# Comparación detección con referencia
tolerance = int(0.05 * fs_ecg)
TP = 0
FP = 0
detected = np.zeros(len(qrs_detections), dtype=bool)

for p in peaks:
    match = False
    for i, qrs in enumerate(qrs_detections):
        if not detected[i] and abs(p - qrs) <= tolerance:
            TP += 1
            detected[i] = True
            match = True
            break
    if not match:
        FP += 1

FN = len(qrs_detections) - TP
sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0
precision = TP / (TP + FP) if (TP + FP) > 0 else 0

#  Resultados detección 
plt.figure(figsize=(12, 4))
plt.plot(ecg_one_lead, label='ECG original')
plt.plot(filtered_signal / np.max(filtered_signal), label='Filtro adaptado (normalizado)')
plt.plot(peaks, ecg_one_lead[peaks], 'rx', label='Detecciones')
plt.plot(qrs_detections, ecg_one_lead[qrs_detections], 'go', label='QRS referencia')
plt.legend()
plt.title("Detección de latidos con filtro adaptado")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")
plt.tight_layout()
plt.show()
```

```python
# Cálculo del error medio entre señal y filtro adaptado en segmento
mean_error = np.mean(filtered_segment - ecg_segment)
print(f"Error medio (filtered_segment - ecg_segment): {mean_error:.5f}")

# Zoom en un segmento para visualización clara
start = int(30 * fs_ecg)
end = int(35 * fs_ecg)

plt.figure(figsize=(12, 4))
plt.plot(ecg_one_lead[start:end], label='ECG original')
plt.plot(filtered_signal[start:end] / np.max(filtered_signal), label='Filtro adaptado (normalizado)')
plt.plot(peaks[(peaks >= start) & (peaks < end)] - start, 
         ecg_one_lead[peaks[(peaks >= start) & (peaks < end)]], 'rx', label='Detecciones')
plt.legend()
plt.title("Zoom: Detección de latidos con filtro adaptado")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Análisis  

## Filtro de mediana e interpolación por Spline Cúbicos

### a. Valores de según el método de filtrado por Mediana o Spline Cúbico

Se parte de una señal $s(n)$ afectada por la línea de base, de la cual se obtiene una versión filtrada: $\hat{x}(n) = s(n) - \hat{b}(n)$ donde $\hat{b}(n)$ representa la estimación de la línea base.

- **s (Original):** Señal ECG sin filtrar.
- **x̂ Filtrada (Mediana):** Señal filtrada con filtro de mediana, que generalmente elimina picos y ruido impulsivo.
- **x̂ Filtrada (Spline):** Señal filtrada con spline cúbico, que ajusta suavemente una línea base usando interpolación.
- **b̂ Línea base (Mediana):** Línea base estimada con método de mediana.
- **b̂ Línea base (Spline):** Línea base estimada con interpolación spline.
  
En la tabla analizada, se observa que al aplicar el **filtro de mediana**, la estimación $\hat{b}(n)$ resulta prácticamente idéntica a la señal original $s(n)$ en todos los puntos del segmento considerado. Como consecuencia, la señal filtrada $\hat{x}(n)$ tiende a cero en toda la ventana analizada.

Esto indica una **eliminación completa del contenido de la señal**, lo cual puede ser deseable en zonas fisiológicamente planas —como el segmento PQ—, pero representa un riesgo en regiones con actividad eléctrica relevante, donde podrían perderse componentes de interés.

En contraste, al aplicar **interpolación por spline cúbico**, se obtiene una línea de base constante (alrededor de $-1 mV$, que no se ajusta punto a punto a $s(n)$, sino que **captura la tendencia general del segmento**.

La señal filtrada resultante conserva amplitudes significativas y variaciones que podrían contener información fisiológica. Este comportamiento refleja que el spline **suaviza sin eliminar abruptamente la señal**, proporcionando una estimación más balanceada de la línea base.

---

### b.Análisis visual de los resultados: segmentos PQ y QRS

Desde el punto de vista visual, las gráficas comparativas permiten observar claramente los efectos estructurales de cada técnica.

#### Segmento PQ

La estimación de la línea de base mediante **spline cúbico** presenta menores residuos en el segmento PQ en comparación con el filtro de mediana. Este resultado es coherente con las propiedades del spline: se trata de un método de interpolación suave, ajustado exactamente a puntos de anclaje fisiológicos —en este caso, el PQ, considerado isoeléctrico—.

Además, el **spline cúbico natural garantiza continuidad hasta la segunda derivada**, lo que lo hace ideal para modelar oscilaciones lentas sin introducir distorsiones en regiones de baja actividad eléctrica.

Por su parte, el filtro de mediana mostró **ligeras sobreestimaciones o subestimaciones** en algunos puntos del segmento PQ. Esto se explica por su naturaleza de **filtro no lineal basado en ventanas móviles**, que no considera la estructura fisiológica de la señal.

Si bien es bueno frente al ruido y útil para estimar niveles de fondo, su **falta de continuidad derivada** y su dependencia del tamaño de la ventana pueden introducir inconsistencias sutiles.

Este tipo de comportamiento es típico de filtros de mediana con ventanas amplias, que no definen una respuesta en frecuencia clara y pueden generar discontinuidades (escalones) en la señal procesada.

#### Segmento QRS

En el complejo QRS, ambos métodos muestran diferencias significativas, pero en sentido opuesto. En este caso, el **filtro de mediana** se desempeña mejor, ya que tiene la capacidad de **preservar bordes abruptos** sin suavizarlos.

Esto es clave para mantener la morfología del complejo QRS, caracterizado por transiciones rápidas y picos pronunciados. La mediana no realiza un suavizado agresivo, por lo que **conserva los detalles estructurales de la señal**.

En cambio, el spline tiende a **suavizar excesivamente estas pendientes pronunciadas**, lo que da lugar a una representación más redondeada del complejo QRS.

Este efecto es una consecuencia directa de su continuidad derivada: si bien es útil para estimar la línea de base en regiones suaves, **no es el método ideal para representar eventos de alta velocidad** como los complejos QRS.

Esta diferencia es crítica al evaluar señales donde se requiere preservar la forma original con precisión.

## 3) Filtro adaptado (matched filter)

### a) Concepto y aplicación

Un filtro adaptado o *matched filter* es un sistema que maximiza la relación señal-ruido (SNR) para detectar una señal conocida en presencia de ruido. Matemáticamente, consiste en convolucionar la señal observada con una versión invertida en el tiempo del patrón que se desea detectar.

En este caso, el patrón `qrs_pattern` representa la forma típica de un complejo QRS en un ECG. Al aplicar el filtro adaptado a la señal ECG, se genera una señal de salida donde los latidos (complejos QRS) producen picos destacados que facilitan su detección automática mediante la búsqueda de máximos.

**Utilidad:**

- Permite detectar latidos cardíacos en señales ECG ruidosas, mejorando la sensibilidad del detector.
- Es especialmente efectivo cuando la forma del QRS es consistente y conocida, como en registros con poca variabilidad morfológica.

**Limitaciones:**

- La variabilidad en la morfología del QRS (causada por arritmias o condiciones cardíacas) puede reducir la eficacia del filtro, ya que el patrón utilizado puede no coincidir con todos los latidos.
- El filtro es sensible a la alineación temporal; si el latido real está desfasado respecto al patrón, la respuesta se debilita.
- Se requiere un preprocesamiento adecuado de la señal, como la normalización y eliminación de tendencias, para que el filtro funcione correctamente.


### b) Detección de latidos y evaluación de performance

Se aplicó el filtro adaptado a la señal ECG y, mediante la detección de picos en la salida filtrada, se identificaron los latidos potenciales. Para evitar detecciones múltiples del mismo latido, se utilizó un intervalo refractario y un umbral adecuado.

Las detecciones se compararon con los latidos de referencia almacenados en `qrs_detections`, considerando una ventana de tolerancia de ±50 ms para definir coincidencias.

Las métricas obtenidas fueron:

- **Verdaderos Positivos (TP):** 1147 latidos detectados correctamente.
- **Falsos Positivos (FP):** 0 detecciones incorrectas.
- **Falsos Negativos (FN):** 756 latidos no detectados.
- **Sensibilidad (Recall):** 60.27%, lo que indica que se detectó aproximadamente el 60% de los latidos reales.
- **Valor Predictivo Positivo (Precision):** 100%, señalando que todas las detecciones realizadas correspondieron efectivamente a latidos (no hubo falsos positivos).

Además, se observó un error medio entre el segmento filtrado y la señal original cercano a -0.024, lo que indica una buena correspondencia en la forma general del patrón detectado.

Estos resultados muestran que el filtro adaptado es bastante estricto, ya que no detectó falsos positivos (FP = 0) y por eso la precisión es muy alta. Esto está bueno porque no se generan falsas alarmas. Pero, por otro lado, la sensibilidad es baja, lo que quiere decir que no logró detectar todos los latidos reales. Esto puede ser porque algunos latidos no tienen exactamente la misma forma que el patrón que usamos, o porque el umbral para decidir si hay un latido fue muy alto.

Para mejorar el rendimiento del detector, se podría probar cambiando el umbral para captar más latidos o usar varios filtros adaptados diferentes para distintos tipos de latidos. También ayudaría hacer un mejor preprocesamiento de la señal para que los latidos estén mejor alineados y normalizados antes de aplicar el filtro.

# Conclusión

El análisis comparativo entre el filtro de mediana y la interpolación por spline cúbico para la estimación de la línea de base en señales de ECG me permitió comprender mejor cómo cada método responde de manera diferente según la naturaleza de la señal y el segmento analizado. 

El filtro de mediana demostró ser especialmente efectivo frente a ruido impulsivo y a distorciones abruptas. Esto se debe a que es un filtro no lineal que preserva bordes y transiciones bruscas, lo que lo hace muy útil para mantener la morfología original de eventos rápidos como el complejo QRS. Esta capacidad para conservar los picos y las formas características en este caso del ECG. No obstante, esta misma propiedad puede ser una limitación en segmentos de la señal que no son planos ni tienen cambios bruscos, ya que el filtro de mediana puede eliminar no solo el ruido sino también componentes útiles y suaves de la señal, lo que puede afectar la calidad general del procesamiento.

En contraste, la interpolación por spline cúbico mostró una mayor capacidad para capturar tendencias lentas y oscilaciones suaves en la línea de base del ECG, ofreciendo una estimación continua y coherente desde el punto de vista fisiológico. Al ajustarse a puntos isoeléctricos como el segmento PQ, esta técnica logra eliminar la deriva de la línea de base sin modificar la señal de interés. La principal desventaja del spline aparece en las regiones con transiciones rápidas, como el QRS, donde la suavidad del método tiende a redondear y atenuar los picos, modificando la morfología original y perdiendo información que podría ser crucial para la detección y análisis de latidos.

Estas diferencias también se reflejaron en los resultados numéricos obtenidos. Por ejemplo, el filtro de mediana puede llegar a anular completamente la señal cuando la línea base estimada es muy similar a la señal original, mientras que el spline conserva variaciones fisiológicas relevantes al no hacer un ajuste punto a punto, sino al captar la tendencia general. Las observaciones visuales acompañaron estos resultados: en los segmentos PQ, el spline mostró una mejor adaptación y menor cantidad de residuos, mientras que en los segmentos QRS, el filtro de mediana logró preservar con mayor fidelidad la forma de los complejos eléctricos, frente a la suavización excesiva que presenta el spline.

En conclusión, la elección entre ambos métodos depende mucho del tipo de señal que se procese y del objetivo final del análisis. La interpolación por spline cúbico es una técnica adecuada para análisis  donde se busca una línea de base fisiológicamente  y se cuenta con la señal completa para un procesamiento más detallado. Por su parte, el filtro de mediana es una solución rápida, simple y eficiente para eliminar interferencias, sobre todo cuando es fundamental preservar la forma de eventos eléctricos rápidos y bruscos. 

---

En cuanto al filtro adaptado (matched filter), funciona como un detector que busca en la señal una forma específica que ya conocemos — en este caso, el patrón típico del complejo QRS. La idea principal del filtro adaptado es maximizar la relación señal-ruido, realzando aquellas partes de la señal que coinciden con el patrón y reduciendo el impacto del ruido y otras interferencias. Esto facilita la detección de latidos en señales que pueden estar muy contaminadas o tener mucha variabilidad.

Sin embargo, noté que el filtro matched es bastante conservador y estricto, lo que se reflejó en que no se detectaron falsos positivos, es decir, que casi ninguna detección fue errónea. Esto es muy bueno para evitar falsas alarmas, Pero esta ventaja tiene un costo: la sensibilidad del detector fue limitada, lo que significa que no todos los latidos reales fueron detectados. Esto puede deberse a que la forma del QRS en la señal real no coincide perfectamente con el patrón usado para el filtro, o a que el umbral para decidir detección fue muy alto, haciendo que algunos latidos menos claros quedaran fuera.

Además, para que el filtro adaptado funcione correctamente, es necesario un preprocesamiento cuidadoso de la señal, como alinear temporalmente el patrón con los latidos y normalizar la amplitud, ya que si la señal y el patrón no están bien ajustados, el filtro puede perder eficacia. Por eso, aunque el matched filter es una herramienta poderosa para detectar eventos con forma conocida en señales ruidosas, también requiere ajustes finos y puede beneficiarse de complementarse con otros métodos que ayuden a captar mayor variabilidad y reducir la cantidad de latidos no detectados.

En definitiva, esta experiencia me permitió entender que el filtrado digital no es solo aplicar un método y listo, sino que se trata de balancear entre detectar bien y evitar falsas alarmas, ajustando parámetros y combinando técnicas para lograr el mejor resultado posible según el contexto y el objetivo del análisis. Aprender a usar y evaluar filtros como el matched filter es clave para poder diseñar sistemas de detección más precisos y confiables.

## a) Propuesta de mejora para la estimación de la línea base usando técnicas multirate con wavelets

Una mejora interesante para los estimadores de la señal de línea base $b$ es aplicar técnicas multirate basadas en la **transformada wavelet**. Este enfoque permite analizar la señal ECG en diferentes escalas temporales y frecuenciales, facilitando la separación de componentes lentos (como la línea base o deriva) y componentes rápidos (como el complejo QRS y ruido).

El método consiste en:

- **Descomposición wavelet multirate:** se aplica una descomposición en varios niveles de la señal ECG utilizando wavelets, que actúa como un análisis multiresolución. Cada nivel corresponde a una versión de la señal con diferente tasa de muestreo (multirate) y banda de frecuencia.

- **Filtrado y eliminación de componentes de baja frecuencia:** se identifican y extraen las aproximaciones en niveles bajos (componentes de baja frecuencia) que contienen la línea base y la deriva. Al eliminar o ajustar estos coeficientes se puede corregir la línea base sin afectar las partes rápidas y útiles de la señal.

- **Reconstrucción de la señal filtrada:** luego de modificar las aproximaciones, se reconstruye la señal mediante la transformada inversa wavelet, recuperando una versión del ECG corregida y con la línea base estimada y eliminada.

Una de las principales ventajas del método wavelet es que utiliza ventanas de ajuste variable en el tiempo y frecuencia. Esto es especialmente útil para una señal como el ECG, que tiene una banda de frecuencia muy amplia, desde componentes muy lentos (menos de 0.5 Hz) hasta frecuencias alrededor de 35 Hz. Gracias a estas ventanas adaptativas, la transformada wavelet permite obtener una mejor aproximación de la línea base en cada rango de frecuencia sin afectar las partes rápidas de la señal, mejorando la precisión del estimador.

Este método tiene varias ventajas adicionales:

- Permite separar la señal en distintas escalas, logrando un filtrado más adaptativo y preciso según la frecuencia de interés.
- Preserva mejor las características importantes del ECG, como picos y bordes, evitando distorsiones típicas de los filtros clásicos lineales.
- Es inherentemente multirate, ya que cada nivel wavelet trabaja con diferentes frecuencias de muestreo, reduciendo la complejidad computacional.
- Puede manejar señales no estacionarias y con artefactos variables, mejorando la robustez del estimador.

El desafío, es que es necesario elegir adecuadamente la familia wavelet y, además, el proceso puede ser más complejo de implementar que técnicas tradicionales.



## b) ¿Podría funcionar el detector basado en el filtro adaptado con un latido perteneciente a otro registro o paciente?

No, como vimos durante el trabajo y en la conclusión, el detector basado en el filtro adaptado está diseñado específicamente para detectar un patrón conocido, en este caso el qrs_pattern que corresponde a un latido particular del registro analizado. Esto explica por qué el filtro logró evitar falsas detecciones (FP = 0) y mantuvo una precisión del 100%, lo que indica que todas las detecciones que hizo correspondieron realmente a latidos.

Sin embargo, también observamos que la sensibilidad fue limitada, con un 60.27%, lo que significa que se perdieron un 39.73% de los latidos reales (756 falsos negativos). Esta baja sensibilidad refleja que el filtro tiene dificultades para detectar latidos que no coinciden exactamente con el patrón utilizado, ya sea por variabilidad morfológica natural, presencia de arritmias o diferencias entre pacientes.

Por lo tanto, si se aplicara este filtro adaptado a latidos provenientes de otro registro o paciente, donde las formas del QRS puedan variar más, es probable que la sensibilidad baje aún más, disminuyendo la capacidad del detector para reconocer esos latidos y aumentando los falsos negativos.

Para superar esta limitación, sería necesario ajustar el umbral del detector, implementar varios filtros adaptados para cubrir diferentes morfologías de latidos, o incluir etapas de preprocesamiento que normalicen y alineen mejor la señal. Alternativamente, métodos más avanzados como aprendizaje automático podrían ayudar a generalizar la detección entre distintos pacientes y condiciones.

En resumen, el filtro adaptado mostró ser muy eficaz para detectar latidos similares al patrón usado (precisión del 100% y FP = 0), pero tiene una sensibilidad limitada (60.27%) que refleja su poca capacidad para generalizar. Esto queda claro en los resultados y debe considerarse al aplicar este tipo de filtro en escenarios clínicos con señales más variables.
