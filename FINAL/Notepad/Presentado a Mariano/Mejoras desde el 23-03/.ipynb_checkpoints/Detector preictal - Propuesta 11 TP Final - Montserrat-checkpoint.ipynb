## Resumen

En este trabajo se desarrolla un detector preictal de crisis epilépticas sobre registros del dataset **CHB-MIT**. El procesamiento se realiza sobre un canal EEG seleccionado del paciente `chb04`, a partir de una cadena compuesta por lectura del *summary*, filtrado pasabanda, remoción de *baseline*, construcción de plantillas preictales y aplicación de un filtro adaptado.  

Se estudian tres estrategias de construcción de plantilla: una plantilla promedio de segmentos preictales alineados, una plantilla manual tomada de una crisis particular y una plantilla idealizada. Para cada una de ellas se comparan dos criterios de umbral: uno basado en la estadística global de la salida del filtro y otro relativo al máximo global de correlación.  

Los resultados muestran que la plantilla promedio alineada ofrece el mejor compromiso entre representatividad morfológica y desempeño del detector, mientras que la plantilla manual resulta más específica pero menos generalizable, y la plantilla idealizada no logra describir adecuadamente la morfología preictal observada en el paciente analizado.


## Estructura del desarrollo

Se organiza el trabajo en las siguientes etapas:

1. **Datos y preprocesamiento**  
   Lectura del *summary*, diseño del filtro, lectura de EDF y remoción de *baseline*.

2. **Exploración temporal y espectral**  
   Inspección visual de crisis y análisis espectral mediante Welch.

3. **Construcción de plantillas**  
   Desarrollo de tres alternativas de plantilla preictal: promedio alineado, manual e idealizada.

4. **Filtro adaptado y detección**

5. **Resultados y discusión final**  
   Comparación de desempeño entre opciones y síntesis de conclusiones.

El trabajo se plantea como un estudio **paciente-específico** y de carácter exploratorio, centrado en la interpretación de la señal y en la comparación entre distintas decisiones de diseño.


### Imports y configuración general

En este primer bloque se cargan las librerías necesarias para todo el desarrollo y se fijan los parámetros generales del trabajo. También se dejan definidos nombres de archivos, rutas, canal de análisis y variables comunes que luego se reutilizan en las distintas etapas del procesamiento.


```python
# IMPORTS

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
from pathlib import Path
import mne
import re
from pytc2.sistemas_lineales import plot_plantilla
from scipy.signal import find_peaks

plt.close("all")

# ARCHIVOS Y LECTURA DEL SUMMARY

# Rutas del dataset
ruta_base = Path(r"C:/Users/monse/OneDrive/Documentos/GitHub/Montserrat-Coto-APS/FINAL")
ruta_summary = ruta_base / "chb04-summary.txt"

# Archivos a procesar
lista_edf = [
    "chb04_01.edf",
    "chb04_05.edf",
    "chb04_08.edf",
    "chb04_28.edf",
]

# Canal elegido
# Ranking de mejores canales
canal_sel = "CZ-PZ"
#canal_sel = "P7-T7"
#canal_sel = "FT9-FT10"
#canal_sel = "T7-FT9"
#canal_sel = "F7-T7"
#canal_sel = "FP2-F8"

# Frecuencia de muestreo esperada
fs_esperada = 256.0

# Diccionarios / listas donde se guardan resultados
eeg_por_archivo = {}
segmentos_previos = []
crisis_filtradas = []
crisis_por_archivo = {}
```

### Lectura del *summary*

En esta etapa se lee el archivo `summary` del paciente, ya que allí se encuentran las referencias temporales de cada crisis. Para cada archivo *.edf*, el *summary* indica si el registro contiene crisis y, en caso afirmativo, los instantes de inicio y fin correspondientes.

Esta información resulta clave para construir ventanas preictales, analizar el comportamiento de la señal alrededor del onset y evaluar posteriormente el desempeño del detector.


```python

if not ruta_summary.exists():
    print("AVISO: no se encontró el summary")
else:
    patron_archivo = re.compile(r"^\s*File Name:\s*(.+?)\s*$", re.IGNORECASE)
    patron_ini = re.compile(r"^\s*Seizure(?:\s+\d+)?\s+Start\s+Time:\s*(\d+)\s*seconds", re.IGNORECASE)
    patron_fin = re.compile(r"^\s*Seizure(?:\s+\d+)?\s+End\s+Time:\s*(\d+)\s*seconds", re.IGNORECASE)

    nombre_actual = None
    inicio_actual = None

    lineas = ruta_summary.read_text(encoding="utf-8", errors="ignore").splitlines()

    for linea in lineas:
        s = linea.strip()

        m_archivo = patron_archivo.match(s)
        if m_archivo:
            nombre_actual = m_archivo.group(1).strip()
            crisis_por_archivo.setdefault(nombre_actual, [])
            inicio_actual = None
            continue

        if nombre_actual is None:
            continue

        m_ini = patron_ini.match(s)
        if m_ini:
            inicio_actual = int(m_ini.group(1))
            continue

        m_fin = patron_fin.match(s)
        if m_fin and inicio_actual is not None:
            fin_actual = int(m_fin.group(1))
            crisis_por_archivo[nombre_actual].append((inicio_actual, fin_actual))
            inicio_actual = None

print("CRISIS ENCONTRADAS EN EL SUMMARY")
for nombre_edf in lista_edf:
    print(nombre_edf, "->", crisis_por_archivo.get(nombre_edf, []))
```

### Diseño del filtro digital

Como etapa de preprocesamiento se utiliza un filtro pasabanda **IIR Butterworth**. Esta aproximación se elige por su respuesta suave en amplitud dentro de la banda pasante y porque evita ondulaciones marcadas que podrían alterar la morfología general de la señal.

En EEG suele resultar útil trabajar con una banda que conserve las componentes lentas y medias de interés clínico, asociadas de manera aproximada a las bandas **delta** (0.5–4 Hz), **theta** (4–8 Hz), **alpha** (8–13 Hz) y parte de **beta** baja (13–30 Hz), mientras que las componentes muy lentas suelen estar dominadas por deriva de línea de base y las más altas pueden contener mayor proporción de ruido o artefactos.

A partir del análisis espectral con Welch realizado más adelante, se observa que la mayor parte del contenido energético relevante de las ventanas preictal, ictal y postictal se concentra por debajo de aproximadamente **25–30 Hz**. Por ese motivo se adopta finalmente una banda de paso de **0.5 a 22 Hz**, con banda de detención por debajo de **0.2 Hz** y por encima de **27 Hz**.

Las especificaciones elegidas son:

- banda de paso: **0.5 a 22 Hz**
- banda de detención: por debajo de **0.2 Hz** y por encima de **27 Hz**
- rizado máximo en banda de paso: **0.5 dB**
- atenuación mínima en banda de detención: **40 dB**

Con este criterio se busca, por un lado, atenuar deriva lenta y componentes muy bajas asociadas al *baseline* y, por otro, concentrar el análisis en una banda útil para EEG que conserve la morfología de interés antes de la crisis.


```python

#DISEÑO DEL FILTRO

filter_type = 'bandpass'
aprox_name = 'butter'

fpass = np.array([0.5, 22.0])      # Banda de paso
fstop = np.array([0.2, 27.0])      # Banda de detención
ripple = 0.5                       # Rizado en banda de paso [dB]
attenuation = 40                   # Atenuación en banda de detención [dB]

mi_sos_butter = sig.iirdesign(
    wp=fpass,
    ws=fstop,
    gpass=ripple,
    gstop=attenuation,
    ftype=aprox_name,
    output='sos',
    fs=fs_esperada
)

# Visualización de plantilla y respuesta en frecuencia
nyq_frec = fs_esperada / 2.0

f_low = np.linspace(0.01, 0.4, 300)
f_fine = np.linspace(0.4, 0.6, 500)
f_high = np.linspace(0.6, nyq_frec, 700)
f_total = np.concatenate((f_low, f_fine, f_high))

w_rad = f_total / nyq_frec * np.pi
w, hh = sig.sosfreqz(mi_sos_butter, worN=w_rad)

plt.figure()
plt.plot(
    w / np.pi * nyq_frec,
    20 * np.log10(np.abs(hh) + 1e-15),
    label='Respuesta del filtro (Butter)'
)
plt.title('Plantilla del filtro digital para EEG (Butter)')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud [dB]')
plt.grid(True)
plt.ylim([-80, 5])

plot_plantilla(
    filter_type=filter_type,
    fpass=fpass,
    ripple=ripple,
    fstop=fstop,
    attenuation=attenuation,
    fs=fs_esperada
)

plt.show()
```

### Lectura de EDF y aplicación del filtro

En este paso se abre cada archivo EDF, se selecciona el canal de interés y se convierte la señal a **microvoltios** para trabajar en la escala estandar. Luego se aplica el filtro pasabanda diseñado previamente mediante un esquema de **fase nula**, con el objetivo de evitar desplazamientos temporales artificiales sobre la morfología de la señal.


```python
#%% APLICACIÓN DEL FILTRO A LA SEÑAL

for nombre_edf in lista_edf:
    ruta_edf = ruta_base / nombre_edf

    if not ruta_edf.exists():
        continue

    try:
        raw = mne.io.read_raw_edf(str(ruta_edf), preload=True, verbose=False)
    except Exception:
        continue

    if canal_sel not in raw.ch_names:
        continue

    data, tiempos_s = raw.get_data(picks=[canal_sel], return_times=True)

    x_V = data[0]
    fs = float(raw.info["sfreq"])

    x_crudo = x_V * 1e6

    print("Canal leído", canal_sel)
    print("Cantidad de muestras:", len(x_crudo))
    print("fs detectada:", fs)

    x_filtrada = sig.sosfiltfilt(mi_sos_butter, x_crudo)

    eeg_por_archivo[nombre_edf] = {
        "x_crudo": x_crudo,
        "x_filtrada": x_filtrada,
        "t": tiempos_s,
        "fs": fs,
        "crisis": crisis_por_archivo.get(nombre_edf, [])
    }
```

```python
#%% GRAFICOS POR ARCHIVO - SEÑAL CRUDA VS FILTRADA

# Acá se grafica un tramo grande de cada archivo.
# Criterio elegido:
# - si el archivo tiene crisis, se toma una ventana de 5 minutos  alrededor de la primera crisis
# - si no tiene crisis, se grafican los primeros 5 minutos
# En todos los casos se marcan con líneas punteadas el inicio y fin de cada crisis que caiga dentro de la ventana graficada.

duracion_plot_s = 600.0   # 300 5 minutos; 180 3 min, 600 10min

# Lista para guardar los segmentos y después hacer un gráfico conjunto
segmentos_archivos = []

for nombre_edf in eeg_por_archivo:
    print("Graficando archivo:", nombre_edf)

    x_crudo = eeg_por_archivo[nombre_edf]["x_crudo"]
    x_filtrada = eeg_por_archivo[nombre_edf]["x_filtrada"]
    fs = eeg_por_archivo[nombre_edf]["fs"]
    lista_crisis = eeg_por_archivo[nombre_edf]["crisis"]

    # Si hay crisis, centrar la ventana en la primera
    if len(lista_crisis) > 0:
        ini_crisis_ref = lista_crisis[0][0]
        t_ini = ini_crisis_ref - duracion_plot_s / 2.0
        t_fin = ini_crisis_ref + duracion_plot_s / 2.0
    else:
        t_ini = 0.0
        t_fin = duracion_plot_s

    # Ajuste para no salir del archivo
    if t_ini < 0:
        t_ini = 0.0
        t_fin = duracion_plot_s

    tiempo_total_archivo = len(x_crudo) / fs

    if t_fin > tiempo_total_archivo:
        t_fin = tiempo_total_archivo
        t_ini = max(0.0, t_fin - duracion_plot_s)

    i0 = int(np.floor(t_ini * fs))
    i1 = int(np.ceil(t_fin * fs))

    if i0 < 0:
        i0 = 0
    if i1 > len(x_crudo):
        i1 = len(x_crudo)

    x_crudo_seg = x_crudo[i0:i1]
    x_filtrada_seg = x_filtrada[i0:i1]
    t_seg_abs = np.arange(i0, i1) / fs

    plt.figure(figsize=(12, 5))

    plt.plot(t_seg_abs, x_crudo_seg, color="tab:blue", label="Crudo", alpha=0.7)
    plt.plot(t_seg_abs, x_filtrada_seg, color="tab:green", label="Filtrado", alpha=0.5)

    # marcar crisis que caen dentro de la ventana
    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        if t_ini <= ini_crisis <= t_fin or t_ini <= fin_crisis <= t_fin:
            if i == 0:
                plt.axvline(ini_crisis, color="c", linestyle="--", linewidth=1.5, label="Inicio crisis")
                plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")
            else:
                plt.axvline(ini_crisis, color="c", linestyle="--", linewidth=1.5)
                plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5)

    plt.title(nombre_edf + " | Señal cruda vs filtrada")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud [uV]")
    plt.grid(True)
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()

    # guardar segmento para el gráfico conjunto
    segmentos_archivos.append({
        "archivo": nombre_edf,
        "t_seg_abs": t_seg_abs,
        "x_crudo_seg": x_crudo_seg,
        "x_filtrada_seg": x_filtrada_seg,
        "crisis": lista_crisis,
        "t_ini": t_ini,
        "t_fin": t_fin
    })
```

### Filtro no lineal de mediana para quitar *baseline* residual

Después del filtrado pasabanda todavía puede persistir una componente lenta residual. Para reducirla se aplica un filtro no lineal de **mediana** y se lo utiliza como estimación del *baseline*, que luego se sustrae de la señal filtrada.

Esta etapa permite trabajar con una señal centrada y más apropiada para la comparación morfológica posterior mediante plantillas y filtro adaptado.


```python
kernel_mediana = 257   # 257 muestras ~ 1 segundo, valor impar

for nombre_edf in eeg_por_archivo:
    x_filtrada = eeg_por_archivo[nombre_edf]["x_filtrada"]

    baseline_mediana = sig.medfilt(x_filtrada, kernel_mediana)
    x_filtrada_sin_baseline = x_filtrada - baseline_mediana

    eeg_por_archivo[nombre_edf]["baseline_mediana"] = baseline_mediana
    eeg_por_archivo[nombre_edf]["x_filtrada_sin_baseline"] = x_filtrada_sin_baseline
```

```python
# GRAFICO CONJUNTO - MEDIANA EN TODOS LOS ARCHIVOS

if len(eeg_por_archivo) > 0:
    n_archivos = len(eeg_por_archivo)

    plt.figure(figsize=(12, 4 * n_archivos))

    k = 1

    for nombre_edf in eeg_por_archivo:
        x_filtrada = eeg_por_archivo[nombre_edf]["x_filtrada"]
        baseline_mediana = eeg_por_archivo[nombre_edf]["baseline_mediana"]
        x_filtrada_sin_baseline = eeg_por_archivo[nombre_edf]["x_filtrada_sin_baseline"]
        t = eeg_por_archivo[nombre_edf]["t"]
        fs = eeg_por_archivo[nombre_edf]["fs"]

        # mismo tramo para todos: primeros 60 s
        t_ini = 0.0
        t_fin = 60.0

        i0 = int(np.floor(t_ini * fs))
        i1 = int(np.ceil(t_fin * fs))

        if i0 < 0:
            i0 = 0
        if i1 > len(t):
            i1 = len(t)

        plt.subplot(n_archivos, 1, k)

        plt.plot(t[i0:i1], x_filtrada[i0:i1], label='EEG filtrado', alpha=0.7)
        plt.plot(t[i0:i1], baseline_mediana[i0:i1], label='Mediana', alpha=0.9)
        plt.plot(t[i0:i1], x_filtrada_sin_baseline[i0:i1], label='EEG sin baseline', alpha=0.7)

        plt.title(nombre_edf + " | Mediana sobre señal filtrada")
        plt.xlabel("Tiempo [s]")
        plt.ylabel("Amplitud [uV]")
        plt.grid(True)
        plt.legend()

        k = k + 1

    plt.tight_layout()
    plt.show()
else:
    print("No hay archivos para graficar.")
```

### Visualización de crisis y exploración visual

Antes de aplicar una detección automática conviene realizar una inspección visual de la señal. Para cada crisis se construye una ventana temporal compuesta por un tramo previo, el intervalo ictal y un tramo posterior. Esta exploración permite observar la morfología general, verificar el efecto del preprocesamiento y reconocer de manera preliminar si existen patrones útiles en la etapa preictal.


```python
# GRAFICOS DE CADA CRISIS - SOLO SEÑAL FILTRADA Y SIN BASELINE

crisis_filtradas = []

for nombre_edf in eeg_por_archivo:
    x_limpia = eeg_por_archivo[nombre_edf]["x_filtrada_sin_baseline"]
    fs = eeg_por_archivo[nombre_edf]["fs"]
    lista_crisis = eeg_por_archivo[nombre_edf]["crisis"]

    if len(lista_crisis) == 0:
        continue

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        Tc = fin_crisis - ini_crisis

        t_ini_total = ini_crisis - Tc
        t_fin_total = fin_crisis + Tc

        if t_ini_total < 0:
            t_ini_total = 0.0

        i0 = int(np.floor(t_ini_total * fs))
        i1 = int(np.ceil(t_fin_total * fs))

        if i0 < 0:
            i0 = 0
        if i1 > len(x_limpia):
            i1 = len(x_limpia)

        x_limpia_seg = x_limpia[i0:i1]
        t_seg_abs = np.arange(i0, i1) / fs

        t_prev_ini = max(0.0, ini_crisis - Tc)
        t_prev_fin = ini_crisis
        t_dur_ini = ini_crisis
        t_dur_fin = fin_crisis
        t_post_ini = fin_crisis
        t_post_fin = fin_crisis + Tc

        plt.figure(figsize=(12, 5))

        plt.plot(t_seg_abs, x_limpia_seg, color="tab:orange", label="Filtrada sin baseline", alpha=0.8)

        plt.axvspan(t_prev_ini, t_prev_fin, color="lightblue", alpha=0.2, label="Previo")
        plt.axvspan(t_dur_ini, t_dur_fin, color="lightcoral", alpha=0.2, label="Durante")
        plt.axvspan(t_post_ini, t_post_fin, color="lightgreen", alpha=0.2, label="Posterior")

        plt.axvline(ini_crisis, color="c", linestyle="--", linewidth=1.5, label="Inicio crisis")
        plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")

        plt.title(nombre_edf + " | Crisis " + str(i + 1) + " | Filtrada y sin baseline")
        plt.xlabel("Tiempo [s]")
        plt.ylabel("Amplitud [uV]")
        plt.grid(True)
        plt.legend(loc="upper right")

        plt.tight_layout()
        plt.show()

        crisis_filtradas.append({
            "archivo": nombre_edf,
            "crisis_num": i + 1,
            "t_seg_abs": t_seg_abs,
            "x_limpia_seg": x_limpia_seg,
            "ini_crisis": ini_crisis,
            "fin_crisis": fin_crisis,
            "Tc": Tc
        })
```

### Análisis espectral con Welch en ventanas preictal, ictal y postictal

Antes de comparar plantillas se realiza una mirada espectral mediante el método de Welch. Esta etapa no se utiliza todavía para decidir el detector, sino para observar si las ventanas preictal, ictal y postictal presentan diferencias de contenido frecuencial que ayuden a interpretar la señal y a revisar si el filtrado pasabanda resulta razonable.

Para cada archivo que contiene crisis se toman tres ventanas:

- una ventana **preictal**, inmediatamente anterior al inicio de crisis;
- la ventana **ictal**, correspondiente al intervalo anotado en el *summary*;
- una ventana **postictal**, inmediatamente posterior al fin de crisis.

En todos los casos se utiliza como duración de referencia la duración de la crisis del archivo. Para los archivos sin crisis se muestra una PSD basal de referencia, de modo de no excluirlos del análisis espectral.

El objetivo principal de esta comparación es verificar dónde se concentra la energía espectral más relevante y decidir si el pasabanda elegido conserva la información útil sin dejar pasar una proporción innecesaria de componentes altas.


```python

#%% ANÁLISIS ESPECTRAL CON WELCH - PREICTAL / ICTAL / POSTICTAL

plt.close("all")

n_archivos = len(eeg_por_archivo)
ncols = 2
nrows = int(np.ceil(n_archivos / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize=(14, 4 * nrows))
axes = np.atleast_1d(axes).ravel()

for ax, nombre_edf in zip(axes, eeg_por_archivo):
    x = eeg_por_archivo[nombre_edf]["x_filtrada_sin_baseline"]
    fs = eeg_por_archivo[nombre_edf]["fs"]
    lista_crisis = eeg_por_archivo[nombre_edf]["crisis"]

    if len(lista_crisis) == 0:
        # Archivo sin crisis: tomo un tramo basal de referencia
        dur_ref_seg = min(10.0, len(x) / fs)
        i0 = 0
        i1 = int(dur_ref_seg * fs)
        x_ref = x[i0:i1]

        if len(x_ref) > 0:
            nperseg = max(64, int(len(x_ref) / 6))

            f_ref, Pxx_ref = sig.welch(
                x_ref,
                fs=fs,
                window='hann',
                nperseg=nperseg,
                noverlap=nperseg // 2,
                detrend=False,
                scaling='density'
            )

            Pxx_ref_db = 10 * np.log10(Pxx_ref + 1e-12)
            Pxx_ref_db -= np.max(Pxx_ref_db)

            ax.plot(f_ref, Pxx_ref_db, linewidth=1.8, label="Basal (referencia)")

        ax.set_title(f"{nombre_edf} - Sin crisis")
    else:
        # Tomo la primera crisis del archivo para comparar preictal, ictal y postictal
        ini_crisis, fin_crisis = lista_crisis[0]
        Tc = fin_crisis - ini_crisis

        t_pre_ini = max(0.0, ini_crisis - Tc)
        t_pre_fin = ini_crisis

        t_ict_ini = ini_crisis
        t_ict_fin = fin_crisis

        t_post_ini = fin_crisis
        t_post_fin = min(len(x) / fs, fin_crisis + Tc)

        segmentos = {
            "Preictal": x[int(t_pre_ini * fs):int(t_pre_fin * fs)],
            "Ictal": x[int(t_ict_ini * fs):int(t_ict_fin * fs)],
            "Postictal": x[int(t_post_ini * fs):int(t_post_fin * fs)],
        }

        colores_welch = {
            "Preictal": "#4C78A8",
            "Ictal": "#B6544B",
            "Postictal": "#6C8E5E"
        }

        for etiqueta in ["Preictal", "Ictal", "Postictal"]:
            x_seg = segmentos[etiqueta]

            if len(x_seg) == 0:
                continue

            nperseg = max(64, int(len(x_seg) / 6))
            if nperseg > len(x_seg):
                nperseg = len(x_seg)

            if nperseg < 16:
                continue

            f_welch, Pxx_welch = sig.welch(
                x_seg,
                fs=fs,
                window='hann',
                nperseg=nperseg,
                noverlap=nperseg // 2,
                detrend=False,
                scaling='density'
            )

            Pxx_welch_db = 10 * np.log10(Pxx_welch + 1e-12)
            Pxx_welch_db -= np.max(Pxx_welch_db)

            ax.plot(
                f_welch,
                Pxx_welch_db,
                linewidth=1.8,
                label=etiqueta,
                color=colores_welch[etiqueta]
            )

        ax.set_title(f"{nombre_edf} - Welch por etapa")

    ax.set_xlabel("Frecuencia [Hz]")
    ax.set_ylabel("PSD [dB/Hz]")
    ax.set_xlim(0, min(40, fs / 2))
    ax.grid(True)
    ax.legend()

for k in range(len(eeg_por_archivo), len(axes)):
    axes[k].axis("off")

plt.suptitle("Comparación espectral con Welch por archivo", y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

### Exploración del patrón candidato y armado de plantilla

Una vez obtenidas señales más limpias, el análisis se concentra en el período **preictal**, ya que el objetivo del trabajo no es detectar el evento ictal completo sino identificar señales previas al onset. A partir de esos segmentos se construyen las distintas plantillas que luego se utilizan en el filtro adaptado.


#### OPCIÓN 1 - PLANTILLA PROMEDIO DE SEGMENTOS PREICTALES ALINEADOS

Primero voy a reconstruir una plantilla promedio a partir de señales reales.  
La ventaja de esta opción es que parte directamente de los datos del paciente y no de una forma inventada.

Sin embargo, para que el promedio tenga sentido, antes debo alinear los segmentos entre sí.  
Si un pico importante aparece en tiempos relativos distintos según la crisis, el promedio directo puede cancelar o suavizar demasiado la forma. En cambio, si logro alinear mejor esos segmentos, espero obtener una plantilla más representativa.

```python
#%% ETAPA 1 - EXPLORACIÓN DEL PATRÓN CANDIDATO

t_antes = 3.0
t_despues = 3.0

segmentos_onset = []

for nombre_edf in eeg_por_archivo:
    x_limpia = eeg_por_archivo[nombre_edf]["x_filtrada_sin_baseline"]
    fs = eeg_por_archivo[nombre_edf]["fs"]
    lista_crisis = eeg_por_archivo[nombre_edf]["crisis"]

    if len(lista_crisis) == 0:
        continue

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        t_ini = ini_crisis - t_antes
        t_fin = ini_crisis + t_despues

        if t_ini < 0:
            t_ini = 0.0

        i0 = int(np.floor(t_ini * fs))
        i1 = int(np.ceil(t_fin * fs))

        if i0 < 0:
            i0 = 0
        if i1 > len(x_limpia):
            i1 = len(x_limpia)

        x_seg = x_limpia[i0:i1]
        t_rel = np.arange(len(x_seg)) / fs - t_antes

        segmentos_onset.append({
            "archivo": nombre_edf,
            "crisis_num": i + 1,
            "fs": fs,
            "x_seg": x_seg,
            "t_rel": t_rel,
            "ini_crisis": ini_crisis,
            "fin_crisis": fin_crisis
        })

print("Cantidad de segmentos de onset guardados:", len(segmentos_onset))

plt.figure(figsize=(10, 5))

for k in range(len(segmentos_onset)):
    t_rel = segmentos_onset[k]["t_rel"]
    x_seg = segmentos_onset[k]["x_seg"]
    x_seg_norm = x_seg / (np.std(x_seg) + 1e-12)

    etiqueta = segmentos_onset[k]["archivo"] + " - crisis " + str(segmentos_onset[k]["crisis_num"])
    plt.plot(t_rel, x_seg_norm, alpha=0.60, label=etiqueta)

plt.axvline(0, color="r", linestyle="--", linewidth=1.5, label="Inicio crisis")
plt.title("Segmentos alrededor del onset, antes de recortar el patrón")
plt.xlabel("Tiempo relativo al inicio [s]")
plt.ylabel("Amplitud normalizada")
plt.grid(True)
plt.legend(loc="center left", bbox_to_anchor=(1.02, 0.5))
plt.tight_layout()
plt.show()
```

```python
#%% ETAPA 1.2 - DEFINIR VENTANA FIJA PREICTAL PARA LA PLANTILLA

t_pat_ini = -2.5
t_pat_fin = 0.0

segmentos_patron = []

for k in range(len(segmentos_onset)):
    t_rel = segmentos_onset[k]["t_rel"]
    x_seg = segmentos_onset[k]["x_seg"]

    mask = (t_rel >= t_pat_ini) & (t_rel <= t_pat_fin)

    t_pat = t_rel[mask]
    x_pat = x_seg[mask]

    segmentos_patron.append({
        "archivo": segmentos_onset[k]["archivo"],
        "crisis_num": segmentos_onset[k]["crisis_num"],
        "t_pat": t_pat,
        "x_pat": x_pat
    })

print("Cantidad de segmentos candidatos para plantilla:", len(segmentos_patron))


# Funciones auxiliares para alinear y evaluar
def normalizar_segmento(x):
    x = np.asarray(x)
    return (x - np.mean(x)) / (np.std(x) + 1e-12)


def desplazar_con_ceros(x, desplazamiento):
    x = np.asarray(x)
    y = np.zeros_like(x)

    if desplazamiento == 0:
        y[:] = x
    elif desplazamiento > 0:
        y[desplazamiento:] = x[:-desplazamiento]
    else:
        d = abs(desplazamiento)
        y[:-d] = x[d:]

    return y


def salida_filtro_adaptado_normalizada(x, plantilla):
    x = np.asarray(x)
    plantilla = np.asarray(plantilla)

    senal_centrada = x - np.mean(x)
    plantilla_centrada = plantilla - np.mean(plantilla)

    correlacion = np.convolve(senal_centrada, plantilla_centrada[::-1], mode="same")

    energia_local = np.sqrt(
        np.convolve(senal_centrada ** 2, np.ones(len(plantilla_centrada)), mode="same")
    ) + 1e-12

    norma_plantilla = np.linalg.norm(plantilla_centrada) + 1e-12

    salida_norm = correlacion / (energia_local * norma_plantilla)

    return salida_norm
```

```python
#%% OPCIÓN 1 - PLANTILLA PROMEDIO DE SEGMENTOS PREICTALES ALINEADOS

largos = []

for k in range(len(segmentos_patron)):
    largos.append(len(segmentos_patron[k]["x_pat"]))

largo_min = min(largos)

# Primero llevo todos los segmentos al mismo largo y los normalizo
matriz_pat_sin_alinear = []

for k in range(len(segmentos_patron)):
    x_pat = segmentos_patron[k]["x_pat"][:largo_min]
    x_pat_norm = normalizar_segmento(x_pat)
    matriz_pat_sin_alinear.append(x_pat_norm)

matriz_pat_sin_alinear = np.array(matriz_pat_sin_alinear)

# Elijo como referencia el segmento que tiene mayor energía media
energias = np.sum(matriz_pat_sin_alinear ** 2, axis=1)
idx_ref = int(np.argmax(energias))
referencia = matriz_pat_sin_alinear[idx_ref, :]

# Alineo cada segmento con la referencia usando correlación cruzada
matriz_pat_alineada = []
lags_aplicados = []

for k in range(matriz_pat_sin_alinear.shape[0]):
    x_pat = matriz_pat_sin_alinear[k, :]

    corr = sig.correlate(x_pat, referencia, mode="full")
    lags = np.arange(-len(referencia) + 1, len(referencia))

    lag_max = lags[np.argmax(corr)]
    x_alineada = desplazar_con_ceros(x_pat, -lag_max)

    matriz_pat_alineada.append(x_alineada)
    lags_aplicados.append(lag_max)

matriz_pat_alineada = np.array(matriz_pat_alineada)

plantilla_promedio_sin_alinear = np.mean(matriz_pat_sin_alinear, axis=0)
plantilla = np.mean(matriz_pat_alineada, axis=0)
t_plantilla = segmentos_patron[0]["t_pat"][:largo_min]

print("Índice del segmento de referencia:", idx_ref)
print("Lags aplicados:", lags_aplicados)

plt.figure(figsize=(10, 5))

for k in range(matriz_pat_sin_alinear.shape[0]):
    plt.plot(t_plantilla, matriz_pat_sin_alinear[k, :], color="gray", alpha=0.25)

plt.plot(t_plantilla, plantilla_promedio_sin_alinear, color="tab:blue", linewidth=2, label="Promedio sin alinear")
plt.plot(t_plantilla, plantilla, color="red", linewidth=2.5, label="Promedio alineado")
plt.axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
plt.title("Opción 1 - Comparación entre promedio sin alinear y alineado")
plt.xlabel("Tiempo relativo al inicio [s]")
plt.ylabel("Amplitud normalizada")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))

for k in range(matriz_pat_alineada.shape[0]):
    plt.plot(t_plantilla, matriz_pat_alineada[k, :], color="gray", alpha=0.30)

plt.plot(t_plantilla, plantilla, color="red", linewidth=2.5, label="Plantilla promedio alineada")
plt.axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
plt.title("Opción 1 - Segmentos alineados y plantilla final")
plt.xlabel("Tiempo relativo al inicio [s]")
plt.ylabel("Amplitud normalizada")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Opción 1 – Filtro adaptado y análisis de máximos de correlación

Con la plantilla promedio alineada se aplica un **filtro adaptado** sobre la señal filtrada y sin *baseline*. La idea es correlacionar la señal con la plantilla para resaltar los tramos donde ambas presentan una forma parecida.

En esta etapa no interesa solamente contar picos. También se guardan:

- el **máximo global** de la salida del filtro;
- el instante donde ocurre ese máximo;
- y el **máximo de correlación dentro de cada ventana preictal**.

Esto permite evaluar si la mayor similitud con la plantilla aparece realmente cerca del inicio de crisis y no en cualquier otro lugar del registro.


En esta etapa se aplica el filtro adaptado usando la plantilla definida para esta opción. La salida del filtro puede interpretarse como una medida de similitud temporal: cuando la correlación toma valores altos, significa que en ese punto del archivo aparece una forma más parecida a la plantilla.

Además del conteo de picos, también se analizan el **máximo global de correlación** y el **máximo dentro de la ventana preictal**. El máximo global indica cuál fue la mayor similitud encontrada en todo el archivo, mientras que el máximo preictal permite verificar si esa similitud aparece en el intervalo previo al inicio de crisis, que es el que realmente interesa para el detector.

Para comparar criterios de decisión se evalúan dos formas de fijar el umbral de detección:

1. **Umbral basado en media y desvío**  
   En este caso se calcula la media de la salida del filtro, `media_salida`, como estimador del nivel medio de correlación del archivo, y el desvío estándar, `desvio_salida`, como medida de dispersión alrededor de ese valor. Luego se fija el umbral como

    $$
    U = \mu_{\text{salida}} + k \,\sigma_{\text{salida}}
    $$

   donde `k_umbral` controla cuán exigente resulta el criterio.

2. **Umbral relativo al máximo global**  
   En este caso se toma el máximo global de correlación, `max_global`, como referencia del mayor valor alcanzado por la salida del filtro en el archivo, y se fija un umbral como una fracción de ese máximo:

     $$
    U = \alpha \, y_{\max}
    $$

   donde `factor_max_global` define qué proporción del máximo global se exige para aceptar otros picos.

En ambos casos, el umbral de altura se complementa con una **distancia mínima** entre picos y una **prominencia mínima**, para evitar que oscilaciones cercanas o máximos poco representativos sean contados como eventos válidos.


```python
# =========================================================
# OPCIÓN 1 | FILTRO ADAPTADO CON UMBRAL ESTADÍSTICO
# =========================================================

resultados_filtro_op1 = {}

# Parámetros de detección
k_umbral = 3.8
distancia_min_seg = 18
factor_prominencia = 1.8


# Ventana preictal que uso para buscar el máximo antes del inicio de crisis
duracion_preictal_seg = 2.5

# Ventanas de visualización alrededor de cada crisis
zoom_antes_seg = 60
zoom_despues_seg = 20

for nombre_edf in eeg_por_archivo:
    datos = eeg_por_archivo[nombre_edf]

    x_limpia = datos["x_filtrada_sin_baseline"]
    fs = datos["fs"]
    lista_crisis = datos["crisis"]
    t = datos["t"]

    salida_filtro = salida_filtro_adaptado_normalizada(x_limpia, plantilla)

    media_salida = np.mean(salida_filtro)
    desvio_salida = np.std(salida_filtro)

    height_threshold = media_salida + k_umbral * desvio_salida
    distance_threshold = int(distancia_min_seg * fs)
    prominence_threshold = factor_prominencia * desvio_salida

    peaks, propiedades = find_peaks(
        salida_filtro,
        height=height_threshold,
        distance=distance_threshold,
        prominence=prominence_threshold
    )

    max_global = np.max(salida_filtro)
    idx_max_global = int(np.argmax(salida_filtro))
    t_max_global = t[idx_max_global]

    maximos_preictales = []

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        i0 = int(max(0, np.floor((ini_crisis - duracion_preictal_seg) * fs)))
        i1 = int(min(len(salida_filtro), np.ceil(ini_crisis * fs)))

        if i1 > i0:
            idx_max_pre = i0 + int(np.argmax(salida_filtro[i0:i1]))
            max_pre = salida_filtro[idx_max_pre]
            t_max_pre = t[idx_max_pre]
        else:
            max_pre = np.nan
            t_max_pre = np.nan

        maximos_preictales.append({
            "crisis_num": i + 1,
            "max_preictal": max_pre,
            "t_max_preictal": t_max_pre
        })

    resultados_filtro_op1[nombre_edf] = {
        "salida_filtro": salida_filtro,
        "fs": fs,
        "t": t,
        "crisis": lista_crisis,
        "peaks": peaks,
        "propiedades": propiedades,
        "media_salida": media_salida,
        "desvio_salida": desvio_salida,
        "height_threshold": height_threshold,
        "distance_threshold": distance_threshold,
        "prominence_threshold": prominence_threshold,
        "max_global": max_global,
        "t_max_global": t_max_global,
        "maximos_preictales": maximos_preictales
    }

    print("Filtro adaptado aplicado a:", nombre_edf)
    print(nombre_edf, "-> cantidad de picos detectados:", len(peaks))
    print(nombre_edf, "-> máximo global:", max_global)
    print(nombre_edf, "-> tiempo del máximo global [s]:", t_max_global)

    for dato in maximos_preictales:
        print(
            "   crisis", dato["crisis_num"],
            "| máximo preictal =", dato["max_preictal"],
            "| tiempo [s] =", dato["t_max_preictal"]
        )

#%% GRÁFICOS DEL FILTRO ADAPTADO
for nombre_edf in resultados_filtro_op1:
    datos_res = resultados_filtro_op1[nombre_edf]

    t = datos_res["t"]
    salida_filtro = datos_res["salida_filtro"]
    peaks = datos_res["peaks"]
    height_threshold = datos_res["height_threshold"]
    lista_crisis = datos_res["crisis"]
    t_max_global = datos_res["t_max_global"]
    max_global = datos_res["max_global"]

    # -----------------------------
    # Figura general por archivo
    # -----------------------------
    plt.figure(figsize=(12, 4))
    plt.plot(t, salida_filtro, color="tab:blue", linewidth=1, label="Salida filtro adaptado normalizada")
    plt.axhline(height_threshold, color="tab:green", linestyle="--", linewidth=1.5, label="Umbral = media + k·desvío")
    plt.plot(t_max_global, max_global, "mo", label="Máximo global")

    # Para no ensuciar el gráfico, muestro los picos detectados con marcadores chicos
    if len(peaks) > 0:
        plt.plot(t[peaks], salida_filtro[peaks], "rx", markersize=4, label="Picos detectados")

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        if i == 0:
            plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5, label="Inicio crisis")
            plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")
        else:
            plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5)
            plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5)

    plt.title(nombre_edf + " | Opción 1 | Salida del filtro adaptado")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Correlación normalizada")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # -----------------------------
    # Zoom por crisis
    # -----------------------------
    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        idx_zoom = np.where((t >= ini_crisis - zoom_antes_seg) & (t <= fin_crisis + zoom_despues_seg))[0]

        if len(idx_zoom) == 0:
            continue

        plt.figure(figsize=(12, 4))
        plt.plot(t[idx_zoom], salida_filtro[idx_zoom], color="tab:blue", linewidth=1, label="Salida filtro adaptado normalizada")
        plt.axhline(height_threshold, color="tab:green", linestyle="--", linewidth=1.5, label="Umbral")

        peaks_zoom = peaks[(t[peaks] >= t[idx_zoom][0]) & (t[peaks] <= t[idx_zoom][-1])]
        if len(peaks_zoom) > 0:
            plt.plot(t[peaks_zoom], salida_filtro[peaks_zoom], "rx", markersize=5, label="Picos detectados")

        dato_pre = datos_res["maximos_preictales"][i]
        if not np.isnan(dato_pre["max_preictal"]):
            plt.plot(
                dato_pre["t_max_preictal"],
                dato_pre["max_preictal"],
                "ko",
                label="Máximo preictal"
            )

        plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5, label="Inicio crisis")
        plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")

        plt.title(nombre_edf + " | Crisis " + str(i + 1) + " | Zoom alrededor de la crisis")
        plt.xlabel("Tiempo [s]")
        plt.ylabel("Correlación normalizada")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```

### Evaluación del filtro

En esta etapa se mantiene el mismo criterio temporal de validación: una detección se considera correcta si aparece dentro de una ventana inmediatamente anterior al inicio de crisis. Se comparan los dos criterios distintos sobre la misma salida del filtro:

- un umbral basado en **media y desvío**;
- y un umbral **relativo al máximo global**.

De esta manera puede discutirse si conviene más un criterio ligado al comportamiento global de la salida o uno que tome como referencia el pico de correlación más alto del archivo. Como la plantilla, la señal filtrada y la lógica de validación se mantienen iguales, cualquier diferencia en los resultados puede atribuirse directamente a la forma de definir el umbral.


```python
# Como la idea es una detección previa, considero correcta una detección si aparece
# en una ventana inmediatamente anterior al inicio de crisis.

ventana_previa_s = 1.5
ventana_posterior_s = 0.0

# Mantengo estos parámetros para ambos criterios
k_umbral = 3.8
factor_max_global = 0.6

def evaluar_detector_con_peaks(resultados_filtro, modo_umbral, nombre_opcion):
    TP = 0
    FP = 0
    FN = 0
    detalle_metricas = {}

    for nombre_edf in resultados_filtro:
        salida_filtro = resultados_filtro[nombre_edf]["salida_filtro"]
        fs = resultados_filtro[nombre_edf]["fs"]
        lista_crisis = resultados_filtro[nombre_edf]["crisis"]
        desvio_salida = resultados_filtro[nombre_edf]["desvio_salida"]
        media_salida = resultados_filtro[nombre_edf]["media_salida"]
        max_global = resultados_filtro[nombre_edf]["max_global"]

        if modo_umbral == "media_desvio":
            height_threshold = media_salida + k_umbral * desvio_salida
            titulo_umbral = "Umbral basado en media y desvío"
        elif modo_umbral == "max_global":
            height_threshold = factor_max_global * max_global
            titulo_umbral = "Umbral relativo al máximo global"
        else:
            raise ValueError("Modo de umbral no válido")

        distance_threshold = int(distancia_min_seg * fs)
        prominence_threshold = factor_prominencia * desvio_salida

        peaks, propiedades = find_peaks(
            salida_filtro,
            height=height_threshold,
            distance=distance_threshold,
            prominence=prominence_threshold
        )

        t_peaks = peaks / fs
        crisis_detectadas = np.zeros(len(lista_crisis), dtype=bool)

        TP_arch = 0
        FP_arch = 0
        FN_arch = 0

        for tpico in t_peaks:
            match = False

            for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
                if (ini_crisis - ventana_previa_s) <= tpico <= (ini_crisis + ventana_posterior_s):
                    if not crisis_detectadas[i]:
                        TP += 1
                        TP_arch += 1
                        crisis_detectadas[i] = True
                        match = True
                        break

            if not match:
                FP += 1
                FP_arch += 1

        FN_arch = int(np.sum(~crisis_detectadas))
        FN += FN_arch

        detalle_metricas[nombre_edf] = {
            "TP": TP_arch,
            "FP": FP_arch,
            "FN": FN_arch,
            "picos": len(peaks),
            "umbral": height_threshold,
            "titulo_umbral": titulo_umbral
        }

    sensibilidad = TP / (TP + FN) if (TP + FN) > 0 else 0
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0

    print(f"RESULTADOS DEL DETECTOR - {nombre_opcion} - {titulo_umbral}")
    for nombre_edf in detalle_metricas:
        d = detalle_metricas[nombre_edf]
        print(f"{nombre_edf} | TP = {d['TP']} | FP = {d['FP']} | FN = {d['FN']} | picos = {d['picos']}")

    print("\nTotales")
    print("TP =", TP)
    print("FP =", FP)
    print("FN =", FN)
    print("Sensibilidad =", sensibilidad)
    print("Precisión =", precision)
    print()

    resumen = {
        "Plantilla": nombre_opcion,
        "Umbral": titulo_umbral,
        "TP": TP,
        "FP": FP,
        "FN": FN,
        "Sensibilidad": sensibilidad,
        "Precision": precision
    }

    return detalle_metricas, resumen

detalle_metricas_op1_media_desvio, resumen_opcion_1_media_desvio = evaluar_detector_con_peaks(resultados_filtro_op1, "media_desvio", "OPCIÓN 1")
detalle_metricas_op1_max_global, resumen_opcion_1_max_global = evaluar_detector_con_peaks(resultados_filtro_op1, "max_global", "OPCIÓN 1")
```

#### OPCIÓN 2 - PLANTILLA MANUAL

En esta segunda alternativa se toma como plantilla un recorte preictal de una crisis elegida manualmente. La idea es trabajar con una forma real de la señal, evitando el posible suavizado o la distorsión que puede aparecer al promediar crisis distintas.

Esta opción también permite evaluar si el patrón elegido manualmente realmente se parece a algo que vuelve a aparecer en otras crisis, o si solo resulta llamativo en ese caso particular.


```python
archivo_manual = "chb04_05.edf"
crisis_manual = 1

plantilla = None
t_plantilla = None
correlaciones_con_manual = []

# ---------------------------------------------------------
# 1) Construyo la plantilla manual a partir de un solo segmento
# ---------------------------------------------------------
for k in range(len(segmentos_patron)):
    x_pat = normalizar_segmento(segmentos_patron[k]["x_pat"][:largo_min])

    if segmentos_patron[k]["archivo"] == archivo_manual and segmentos_patron[k]["crisis_num"] == crisis_manual:
        t_plantilla = segmentos_patron[k]["t_pat"][:largo_min]
        plantilla = x_pat.copy()

if plantilla is None:
    print("No se encontró la crisis elegida para la plantilla manual.")
else:
    # ---------------------------------------------------------
    # 2) Comparo todos los segmentos reales contra la plantilla manual
    #    Primero sin alinear y después alineados a la plantilla manual
    # ---------------------------------------------------------
    matriz_pat_sin_alinear = []
    matriz_pat_alineada = []
    lags_aplicados = []

    for k in range(len(segmentos_patron)):
        x_pat = normalizar_segmento(segmentos_patron[k]["x_pat"][:largo_min])
        matriz_pat_sin_alinear.append(x_pat)

        corr = sig.correlate(x_pat, plantilla, mode="full")
        lags = np.arange(-len(plantilla) + 1, len(plantilla))
        lag_max = lags[np.argmax(corr)]

        x_alineada = desplazar_con_ceros(x_pat, -lag_max)
        matriz_pat_alineada.append(x_alineada)
        lags_aplicados.append(lag_max)

        corr_sin_alinear = np.corrcoef(plantilla, x_pat)[0, 1]
        corr_alineada = np.corrcoef(plantilla, x_alineada)[0, 1]

        correlaciones_con_manual.append({
            "archivo": segmentos_patron[k]["archivo"],
            "crisis_num": segmentos_patron[k]["crisis_num"],
            "corr_sin_alinear": corr_sin_alinear,
            "corr_alineada": corr_alineada,
            "lag": lag_max
        })

    matriz_pat_sin_alinear = np.array(matriz_pat_sin_alinear)
    matriz_pat_alineada = np.array(matriz_pat_alineada)

    print("Plantilla manual tomada de:", archivo_manual, "| crisis", crisis_manual)
    print("Lags aplicados respecto de la plantilla manual:", lags_aplicados)

    # ---------------------------------------------------------
    # 3) Subplot: comparación visual sin alinear vs alineado
    # ---------------------------------------------------------
    fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

    for k in range(matriz_pat_sin_alinear.shape[0]):
        ax[0].plot(t_plantilla, matriz_pat_sin_alinear[k, :], color="gray", alpha=0.25)

    ax[0].plot(t_plantilla, plantilla, color="tab:blue", linewidth=2.5, label="Plantilla manual")
    ax[0].axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
    ax[0].set_title("Opción 2 - Comparación entre segmentos sin alinear y plantilla manual")
    ax[0].set_ylabel("Amplitud normalizada")
    ax[0].grid(True)
    ax[0].legend()

    for k in range(matriz_pat_alineada.shape[0]):
        ax[1].plot(t_plantilla, matriz_pat_alineada[k, :], color="gray", alpha=0.25)

    ax[1].plot(t_plantilla, plantilla, color="red", linewidth=2.5, label="Plantilla manual")
    ax[1].axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
    ax[1].set_title("Opción 2 - Comparación entre segmentos alineados y plantilla manual")
    ax[1].set_xlabel("Tiempo relativo al inicio [s]")
    ax[1].set_ylabel("Amplitud normalizada")
    ax[1].grid(True)
    ax[1].legend()

    plt.tight_layout()
    plt.show()

    # ---------------------------------------------------------
    # 4) Muestro las correlaciones con los demás segmentos
    # ---------------------------------------------------------
    print("Correlación de la plantilla manual con los demás segmentos:")
    for dato in correlaciones_con_manual:
        print(
            dato["archivo"],
            "| crisis", dato["crisis_num"],
            "| corr sin alinear =", dato["corr_sin_alinear"],
            "| corr alineada =", dato["corr_alineada"],
            "| lag =", dato["lag"]
        )
```

En esta etapa se aplica el filtro adaptado usando la plantilla definida para esta opción. La salida del filtro puede interpretarse como una medida de similitud temporal: cuando la correlación toma valores altos, significa que en ese punto del archivo aparece una forma más parecida a la plantilla.


```python
# =========================================================
# OPCIÓN 2 | FILTRO ADAPTADO CON UMBRAL ESTADÍSTICO
# =========================================================

resultados_filtro_op2 = {}

# Mantengo los mismos parámetros que en la opción 1
k_umbral = 3.8
distancia_min_seg = 18
factor_prominencia = 1.8

# Ventana preictal para buscar el máximo antes del inicio de crisis
duracion_preictal_seg = 2.5

# Ventanas de visualización alrededor de cada crisis
zoom_antes_seg = 60
zoom_despues_seg = 20

for nombre_edf in eeg_por_archivo:
    datos = eeg_por_archivo[nombre_edf]

    x_limpia = datos["x_filtrada_sin_baseline"]
    fs = datos["fs"]
    lista_crisis = datos["crisis"]
    t = datos["t"]

    salida_filtro = salida_filtro_adaptado_normalizada(x_limpia, plantilla)

    media_salida = np.mean(salida_filtro)
    desvio_salida = np.std(salida_filtro)

    height_threshold = media_salida + k_umbral * desvio_salida
    distance_threshold = int(distancia_min_seg * fs)
    prominence_threshold = factor_prominencia * desvio_salida

    peaks, propiedades = find_peaks(
        salida_filtro,
        height=height_threshold,
        distance=distance_threshold,
        prominence=prominence_threshold
    )

    max_global = np.max(salida_filtro)
    idx_max_global = int(np.argmax(salida_filtro))
    t_max_global = t[idx_max_global]

    maximos_preictales = []

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        i0 = int(max(0, np.floor((ini_crisis - duracion_preictal_seg) * fs)))
        i1 = int(min(len(salida_filtro), np.ceil(ini_crisis * fs)))

        if i1 > i0:
            idx_max_pre = i0 + int(np.argmax(salida_filtro[i0:i1]))
            max_pre = salida_filtro[idx_max_pre]
            t_max_pre = t[idx_max_pre]
        else:
            max_pre = np.nan
            t_max_pre = np.nan

        maximos_preictales.append({
            "crisis_num": i + 1,
            "max_preictal": max_pre,
            "t_max_preictal": t_max_pre
        })

    resultados_filtro_op2[nombre_edf] = {
        "salida_filtro": salida_filtro,
        "fs": fs,
        "t": t,
        "crisis": lista_crisis,
        "peaks": peaks,
        "propiedades": propiedades,
        "media_salida": media_salida,
        "desvio_salida": desvio_salida,
        "height_threshold": height_threshold,
        "distance_threshold": distance_threshold,
        "prominence_threshold": prominence_threshold,
        "max_global": max_global,
        "t_max_global": t_max_global,
        "maximos_preictales": maximos_preictales
    }

    print("Filtro adaptado aplicado a:", nombre_edf)
    print(nombre_edf, "-> cantidad de picos detectados:", len(peaks))
    print(nombre_edf, "-> máximo global:", max_global)
    print(nombre_edf, "-> tiempo del máximo global [s]:", t_max_global)

    for dato in maximos_preictales:
        print(
            "   crisis", dato["crisis_num"],
            "| máximo preictal =", dato["max_preictal"],
            "| tiempo [s] =", dato["t_max_preictal"]
        )

#%% GRÁFICOS DEL FILTRO ADAPTADO
for nombre_edf in resultados_filtro_op2:
    datos_res = resultados_filtro_op2[nombre_edf]

    t = datos_res["t"]
    salida_filtro = datos_res["salida_filtro"]
    peaks = datos_res["peaks"]
    height_threshold = datos_res["height_threshold"]
    lista_crisis = datos_res["crisis"]
    t_max_global = datos_res["t_max_global"]
    max_global = datos_res["max_global"]

    # Figura general por archivo
    plt.figure(figsize=(12, 4))
    plt.plot(t, salida_filtro, color="tab:blue", linewidth=1, label="Salida filtro adaptado normalizada")
    plt.axhline(height_threshold, color="tab:green", linestyle="--", linewidth=1.5, label="Umbral = media + k·desvío")
    plt.plot(t_max_global, max_global, "mo", label="Máximo global")

    if len(peaks) > 0:
        plt.plot(t[peaks], salida_filtro[peaks], "rx", markersize=4, label="Picos detectados")

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        if i == 0:
            plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5, label="Inicio crisis")
            plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")
        else:
            plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5)
            plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5)

    plt.title(nombre_edf + " | Opción 2 | Salida del filtro adaptado")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Correlación normalizada")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Zoom alrededor de cada crisis
    for dato_max in datos_res["maximos_preictales"]:
        crisis_num = dato_max["crisis_num"]
        max_pre = dato_max["max_preictal"]
        t_max_pre = dato_max["t_max_preictal"]

        ini_crisis, fin_crisis = lista_crisis[crisis_num - 1]

        idx_zoom = np.where((t >= ini_crisis - zoom_antes_seg) & (t <= fin_crisis + zoom_despues_seg))[0]

        if len(idx_zoom) == 0:
            continue

        plt.figure(figsize=(12, 4))
        plt.plot(t[idx_zoom], salida_filtro[idx_zoom], color="tab:blue", linewidth=1, label="Salida filtro adaptado normalizada")
        plt.axhline(height_threshold, color="tab:green", linestyle="--", linewidth=1.5, label="Umbral")

        peaks_zoom = peaks[(t[peaks] >= t[idx_zoom][0]) & (t[peaks] <= t[idx_zoom][-1])]
        if len(peaks_zoom) > 0:
            plt.plot(t[peaks_zoom], salida_filtro[peaks_zoom], "rx", markersize=5, label="Picos detectados")

        if not np.isnan(t_max_pre):
            plt.plot(t_max_pre, max_pre, "ko", label="Máximo preictal")

        plt.axvline(ini_crisis, color="r", linestyle="--", linewidth=1.5, label="Inicio crisis")
        plt.axvline(fin_crisis, color="m", linestyle="--", linewidth=1.5, label="Fin crisis")

        plt.title(f"{nombre_edf} | Crisis {crisis_num} | Zoom alrededor de la crisis")
        plt.xlabel("Tiempo [s]")
        plt.ylabel("Correlación normalizada")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```

### Evaluación del filtro

#### OPCIÓN 3 - PLANTILLA IDEALIZADA

En esta tercera alternativa se mantiene la misma lógica general de análisis que en las otras opciones, pero se reemplaza la plantilla por una forma **idealizada**. La idea no es copiar exactamente una crisis real, sino construir una señal sintética simple que conserve algunos rasgos observados en los segmentos preictales.

Antes de usar esa plantilla en el filtro adaptado, primero se alinean los segmentos reales entre sí para revisar mejor su morfología y para tener una referencia visual más clara de qué forma conviene idealizar.


```python
#%% OPCIÓN 3 - ALINEACIÓN DE SEGMENTOS PREICTALES REALES

largos = []

for k in range(len(segmentos_patron)):
    largos.append(len(segmentos_patron[k]["x_pat"]))

largo_min = min(largos)

matriz_pat_sin_alinear = []

for k in range(len(segmentos_patron)):
    x_pat = segmentos_patron[k]["x_pat"][:largo_min]
    x_pat_norm = normalizar_segmento(x_pat)
    matriz_pat_sin_alinear.append(x_pat_norm)

matriz_pat_sin_alinear = np.array(matriz_pat_sin_alinear)

energias = np.sum(matriz_pat_sin_alinear ** 2, axis=1)
idx_ref = int(np.argmax(energias))
referencia = matriz_pat_sin_alinear[idx_ref, :]

matriz_pat_alineada = []
lags_aplicados = []

for k in range(matriz_pat_sin_alinear.shape[0]):
    x_pat = matriz_pat_sin_alinear[k, :]

    corr = sig.correlate(x_pat, referencia, mode="full")
    lags = np.arange(-len(referencia) + 1, len(referencia))

    lag_max = lags[np.argmax(corr)]
    x_alineada = desplazar_con_ceros(x_pat, -lag_max)

    matriz_pat_alineada.append(x_alineada)
    lags_aplicados.append(lag_max)

matriz_pat_alineada = np.array(matriz_pat_alineada)

t_plantilla = segmentos_patron[0]["t_pat"][:largo_min]

print("Lags aplicados para alinear los segmentos de la opción 3:", lags_aplicados)

plt.figure(figsize=(10, 5))

for k in range(matriz_pat_alineada.shape[0]):
    plt.plot(t_plantilla, matriz_pat_alineada[k, :], color="gray", alpha=0.30)

plt.axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
plt.title("Opción 3 - Segmentos preictales alineados usados como guía")
plt.xlabel("Tiempo relativo al inicio [s]")
plt.ylabel("Amplitud normalizada")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

A partir de la referencia visual de los segmentos alineados, se define una plantilla sintética simple tipo punta-onda. No se busca una copia exacta de ninguna crisis en particular, sino una forma controlada que concentre la energía en una zona cercana al inicio de crisis y que permita probar si una plantilla menos dependiente del ruido mejora la detección.

Una vez generada, esa plantilla idealizada se compara contra los segmentos alineados para ver si, al menos de manera cualitativa, la forma elegida resulta compatible con los patrones reales.


```python
#%% OPCIÓN 3 - GENERACIÓN DE PLANTILLA IDEALIZADA

def gauss(t, A, mu, sigma):
    return A * np.exp(-0.5 * ((t - mu) / sigma) ** 2)

#plantilla = (
#    gauss(t_plantilla,  1.00, -0.16, 0.035) +
#    gauss(t_plantilla, -0.75, -0.07, 0.030)
#)

plantilla = (
    gauss(t_plantilla, 1.20, -0.105, 0.012) +
    gauss(t_plantilla, -0.50, -0.078, 0.012) +
    gauss(t_plantilla, 0.70, -0.050, 0.010)
)

max_abs = np.max(np.abs(plantilla))
if max_abs > 0:
    plantilla = plantilla / max_abs

escala_real = np.median(np.max(np.abs(matriz_pat_alineada), axis=1))
plantilla = 0.50 * escala_real * plantilla

fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

for k in range(matriz_pat_alineada.shape[0]):
    ax[0].plot(t_plantilla, matriz_pat_alineada[k, :], color="gray", alpha=0.30)

ax[0].axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
ax[0].set_title("Opción 3 - Segmentos alineados")
ax[0].set_ylabel("Amplitud normalizada")
ax[0].grid(True)
ax[0].legend()

for k in range(matriz_pat_alineada.shape[0]):
    ax[1].plot(t_plantilla, matriz_pat_alineada[k, :], color="gray", alpha=0.25)

ax[1].plot(t_plantilla, plantilla, color="red", linewidth=2.5, label="Plantilla idealizada")
ax[1].axvline(0, color="k", linestyle="--", linewidth=1.5, label="Inicio crisis")
ax[1].set_title("Opción 3 - Comparación entre segmentos alineados y plantilla idealizada")
ax[1].set_xlabel("Tiempo relativo al inicio [s]")
ax[1].set_ylabel("Amplitud normalizada")
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# =========================================================
# OPCIÓN 3 | FILTRO ADAPTADO CON UMBRAL ESTADÍSTICO
# =========================================================

resultados_filtro_op3 = {}

# Mantengo los mismos parámetros que en las opciones 1 y 2
k_umbral = 3.8
distancia_min_seg = 18
factor_prominencia = 1.8

# Ventana preictal para buscar el máximo antes del inicio de crisis
duracion_preictal_seg = 2.5

# Ventanas de visualización alrededor de cada crisis
zoom_antes_seg = 60
zoom_despues_seg = 20

for nombre_edf in eeg_por_archivo:
    datos = eeg_por_archivo[nombre_edf]

    x_limpia = datos["x_filtrada_sin_baseline"]
    fs = datos["fs"]
    lista_crisis = datos["crisis"]
    t = datos["t"]

    salida_filtro = salida_filtro_adaptado_normalizada(x_limpia, plantilla)

    media_salida = np.mean(salida_filtro)
    desvio_salida = np.std(salida_filtro)

    height_threshold = media_salida + k_umbral * desvio_salida
    distance_threshold = int(distancia_min_seg * fs)
    prominence_threshold = factor_prominencia * desvio_salida

    peaks, propiedades = find_peaks(
        salida_filtro,
        height=height_threshold,
        distance=distance_threshold,
        prominence=prominence_threshold
    )

    idx_max_global = int(np.argmax(salida_filtro))
    max_global = salida_filtro[idx_max_global]
    t_max_global = t[idx_max_global]

    resultados_filtro_op3[nombre_edf] = {
        "salida_filtro": salida_filtro,
        "peaks": peaks,
        "fs": fs,
        "t": t,
        "crisis": lista_crisis,
        "max_global": max_global,
        "t_max_global": t_max_global,
        "media_salida": media_salida,
        "desvio_salida": desvio_salida,
        "umbral": height_threshold
    }

    print("Filtro adaptado aplicado a:", nombre_edf)
    print(nombre_edf, "-> cantidad de picos detectados:", len(peaks))
    print(nombre_edf, "-> máximo global:", max_global)
    print(nombre_edf, "-> tiempo del máximo global [s]:", t_max_global)

    maximos_preictales = []

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        i0 = int(max(0, np.floor((ini_crisis - duracion_preictal_seg) * fs)))
        i1 = int(min(len(salida_filtro), np.ceil(ini_crisis * fs)))

        if i1 > i0:
            idx_local = i0 + int(np.argmax(salida_filtro[i0:i1]))
            max_pre = salida_filtro[idx_local]
            t_max_pre = t[idx_local]
        else:
            max_pre = np.nan
            t_max_pre = np.nan

        maximos_preictales.append({
            "crisis_num": i + 1,
            "max_preictal": max_pre,
            "t_max_preictal": t_max_pre
        })

        print("   crisis", i + 1, "| máximo preictal =", max_pre, "| tiempo [s] =", t_max_pre)

    resultados_filtro_op3[nombre_edf]["maximos_preictales"] = maximos_preictales

    plt.figure(figsize=(12, 4))
    plt.plot(t, salida_filtro, label="Salida filtro adaptado normalizada", linewidth=1)
    plt.axhline(height_threshold, color="green", linestyle="--", label="Umbral = media + k·desvío")
    plt.plot(t[idx_max_global], salida_filtro[idx_max_global], "mo", label="Máximo global")

    if len(peaks) > 0:
        plt.plot(t[peaks], salida_filtro[peaks], "rx", markersize=5, label="Picos detectados")

    for j, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        if j == 0:
            plt.axvline(ini_crisis, color="red", linestyle="--", label="Inicio crisis")
            plt.axvline(fin_crisis, color="magenta", linestyle="--", label="Fin crisis")
        else:
            plt.axvline(ini_crisis, color="red", linestyle="--")
            plt.axvline(fin_crisis, color="magenta", linestyle="--")

    plt.title(f"{nombre_edf} | Opción 3 | Salida del filtro adaptado")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Correlación normalizada")
    plt.legend()
    plt.grid(True)
    plt.show()

    for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
        idx_zoom = np.where((t >= ini_crisis - zoom_antes_seg) & (t <= fin_crisis + zoom_despues_seg))[0]

        if len(idx_zoom) == 0:
            continue

        plt.figure(figsize=(12, 4))
        plt.plot(t[idx_zoom], salida_filtro[idx_zoom], label="Salida filtro adaptado normalizada", linewidth=1)
        plt.axhline(height_threshold, color="green", linestyle="--", label="Umbral")

        peaks_zoom = peaks[(t[peaks] >= t[idx_zoom][0]) & (t[peaks] <= t[idx_zoom][-1])]
        if len(peaks_zoom) > 0:
            plt.plot(t[peaks_zoom], salida_filtro[peaks_zoom], "rx", markersize=6, label="Picos detectados")

        max_pre = maximos_preictales[i]["max_preictal"]
        t_max_pre = maximos_preictales[i]["t_max_preictal"]
        if not np.isnan(max_pre):
            plt.plot(t_max_pre, max_pre, "ko", label="Máximo preictal")

        plt.axvline(ini_crisis, color="red", linestyle="--", label="Inicio crisis")
        plt.axvline(fin_crisis, color="magenta", linestyle="--", label="Fin crisis")

        plt.title(f"{nombre_edf} | Crisis {i+1} | Zoom alrededor de la crisis")
        plt.xlabel("Tiempo [s]")
        plt.ylabel("Correlación normalizada")
        plt.legend()
        plt.grid(True)
        plt.show()
```

### Evaluación del filtro

En esta etapa se mantiene el mismo criterio temporal de validación: una detección se considera correcta si aparece dentro de una ventana inmediatamente anterior al inicio de crisis. Sin embargo, en lugar de evaluar un único umbral, se comparan **dos criterios distintos sobre la misma salida del filtro**:

- un umbral basado en **media y desvío**;
- y un umbral **relativo al máximo global**.

De esta manera puede discutirse si conviene más un criterio ligado al comportamiento global de la salida o uno que tome como referencia el pico de correlación más alto del archivo. Como la plantilla, la señal filtrada y la lógica de validación se mantienen iguales, cualquier diferencia en los resultados puede atribuirse directamente a la forma de definir el umbral.


```python
# Como la idea es una detección previa, considero correcta una detección si aparece
# en una ventana inmediatamente anterior al inicio de crisis.

ventana_previa_s = 1.5
ventana_posterior_s = 0.0

# Mantengo estos parámetros para ambos criterios
k_umbral = 3.8
factor_max_global = 0.6

def evaluar_detector_con_peaks(resultados_filtro, modo_umbral, nombre_opcion):
    TP = 0
    FP = 0
    FN = 0
    detalle_metricas = {}

    for nombre_edf in resultados_filtro:
        salida_filtro = resultados_filtro[nombre_edf]["salida_filtro"]
        fs = resultados_filtro[nombre_edf]["fs"]
        lista_crisis = resultados_filtro[nombre_edf]["crisis"]
        desvio_salida = resultados_filtro[nombre_edf]["desvio_salida"]
        media_salida = resultados_filtro[nombre_edf]["media_salida"]
        max_global = resultados_filtro[nombre_edf]["max_global"]

        if modo_umbral == "media_desvio":
            height_threshold = media_salida + k_umbral * desvio_salida
            titulo_umbral = "Umbral basado en media y desvío"
        elif modo_umbral == "max_global":
            height_threshold = factor_max_global * max_global
            titulo_umbral = "Umbral relativo al máximo global"
        else:
            raise ValueError("Modo de umbral no válido")

        distance_threshold = int(distancia_min_seg * fs)
        prominence_threshold = factor_prominencia * desvio_salida

        peaks, propiedades = find_peaks(
            salida_filtro,
            height=height_threshold,
            distance=distance_threshold,
            prominence=prominence_threshold
        )

        t_peaks = peaks / fs
        crisis_detectadas = np.zeros(len(lista_crisis), dtype=bool)

        TP_arch = 0
        FP_arch = 0
        FN_arch = 0

        for tpico in t_peaks:
            match = False

            for i, (ini_crisis, fin_crisis) in enumerate(lista_crisis):
                if (ini_crisis - ventana_previa_s) <= tpico <= (ini_crisis + ventana_posterior_s):
                    if not crisis_detectadas[i]:
                        TP += 1
                        TP_arch += 1
                        crisis_detectadas[i] = True
                        match = True
                        break

            if not match:
                FP += 1
                FP_arch += 1

        FN_arch = int(np.sum(~crisis_detectadas))
        FN += FN_arch

        detalle_metricas[nombre_edf] = {
            "TP": TP_arch,
            "FP": FP_arch,
            "FN": FN_arch,
            "picos": len(peaks),
            "umbral": height_threshold,
            "titulo_umbral": titulo_umbral
        }

    sensibilidad = TP / (TP + FN) if (TP + FN) > 0 else 0
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0

    print(f"RESULTADOS DEL DETECTOR - {nombre_opcion} - {titulo_umbral}")
    for nombre_edf in detalle_metricas:
        d = detalle_metricas[nombre_edf]
        print(f"{nombre_edf} | TP = {d['TP']} | FP = {d['FP']} | FN = {d['FN']} | picos = {d['picos']}")

    print("\nTotales")
    print("TP =", TP)
    print("FP =", FP)
    print("FN =", FN)
    print("Sensibilidad =", sensibilidad)
    print("Precisión =", precision)
    print()

    resumen = {
        "Plantilla": nombre_opcion,
        "Umbral": titulo_umbral,
        "TP": TP,
        "FP": FP,
        "FN": FN,
        "Sensibilidad": sensibilidad,
        "Precision": precision
    }

    return detalle_metricas, resumen

detalle_metricas_op3_media_desvio, resumen_opcion_3_media_desvio = evaluar_detector_con_peaks(resultados_filtro_op3, "media_desvio", "OPCIÓN 3")
detalle_metricas_op3_max_global, resumen_opcion_3_max_global = evaluar_detector_con_peaks(resultados_filtro_op3, "max_global", "OPCIÓN 3")
```

Más que buscar un detector definitivo, interesa dejar en claro qué decisión de diseño funcionó mejor, qué limitaciones aparecieron y qué conclusiones pueden extraerse sobre la representatividad de cada plantilla.

- si conviene más un umbral ligado al **comportamiento global** de la salida o uno relativo al **máximo global de correlación**;
- y si los máximos más grandes del filtro adaptado corresponden a patrones preictales reales o a eventos de alta energía no deseados.

### Discusión y conclusiones finales

A partir de la comparación entre las tres opciones, se observa que la **opción 1** fue la que mostró el mejor equilibrio entre detección y representatividad de la plantilla. La construcción de una plantilla promedio a partir de segmentos reales alineados permitió conservar información morfológica del paciente sin depender por completo de una única crisis.

La **opción 2** resultó más específica, pero también más dependiente del segmento elegido manualmente. Esto sugiere que una plantilla tomada de una sola crisis puede describir bien ese episodio particular, pero no necesariamente generalizar al resto de las crisis del mismo paciente.

La **opción 3** fue la menos efectiva. Aunque conceptualmente resultó útil como prueba de referencia, la plantilla idealizada no logró representar adecuadamente la morfología real observada en los segmentos preictales y no produjo detecciones verdaderas.

Respecto de los umbrales, el criterio basado en **media y desvío** fue el que mostró el mejor compromiso global. El criterio relativo al máximo global permitió explorar una alternativa coherente con la idea de tomar la mayor correlación del archivo como referencia, pero en la práctica resultó mucho más sensible a picos extremos y produjo una cantidad mayor de falsas detecciones.

En conjunto, los resultados sugieren que, para este paciente y bajo la estrategia de filtro adaptado utilizada, las plantillas construidas a partir de datos reales son más adecuadas que una forma sintética idealizada. También muestran que la alineación previa de segmentos resulta importante cuando se desea construir una plantilla promedio con cierta estabilidad morfológica.

Como trabajo futuro, sería razonable profundizar en el tratamiento de artefactos, revisar estrategias adicionales de segmentación y explorar criterios complementarios de decisión que no dependan únicamente de la amplitud de los máximos de correlación.
